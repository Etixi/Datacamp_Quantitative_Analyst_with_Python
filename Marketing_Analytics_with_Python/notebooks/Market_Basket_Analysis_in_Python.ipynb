####**6. Utilisation de l’analyse du panier de consommation**
+ **L’analyse du panier de consommation est structurée autour de l’utilisation de ce que l’on appelle des « règles d’association »**.

####**2. Mesures et élagage**

+ **Une métrique est une mesure de la performance d’une règle**. 

+ Par exemple, sous une métrique, la règle « si l’humour alors la poésie » pourrait correspondre au nombre 0-point-81. 

+ La même métrique pourrait donner 0 point 23 pour « si la fiction alors voyage ». 

+ **L’élagage utilise une métrique pour ignorer les règles**. 

+ Par exemple, nous ne pourrions conserver que les règles dont la valeur métrique est supérieure à 0 point 50. 

+ Dans l’exemple que nous avons donné, nous regarderiez « si l’humour puis la poésie » et écarterions « si la fiction puis voyage ».

####**3. La métrique la plus simple**

+ **La métrique la plus simple est appelée support, qui mesure la fréquence à laquelle les ensembles d’éléments apparaissent dans les transactions**. 

+ **La prise en charge peut également être appliquée à des éléments uniques**. 

+ Par exemple, dans l’ensemble de données sur les petites épiceries, le lait est l’un des neuf éléments qui apparaissent dans les transactions. 

+ Nous pouvons calculer le support du lait comme le nombre de transactions qui contiennent du lait, divisé par le nombre total de transactions.

####**4. Prise en charge de la langue**
+ À titre d’exemple concret, vérifions la prise en charge de la langue dans les 10 premières transactions de l’ensemble de données de la librairie. 

+ La langue apparaît dans les transactions 1 et 3. Ainsi, la valeur de support est 2 sur 10 ou 0-point-2.

####**5. Prise en charge de {humour} $\rightarrow$ {langue}**

+ ***Que se passe-t-il si nous voulons plutôt vérifier le soutien à une règle d’association, telle que « si l’humour puis le langage »?***

+ Nous calculerions la part des transactions qui contenaient à la fois de l’humour et du langage. 

+ Dans ce cas, nous pouvons voir qu’il n’y en a qu’un, donc le support est de 1 sur 10 ou 0-point-1. 

+ Notez que nous obtiendrions la même valeur si nous calculions plutôt le soutien pour « si le langage puis l’humour ».

#####**Instructions**

+ Calculez le support pour l'antécédent et NON conséquent.

```python
def conviction(antecedent, consequent):
	# Compute support for antecedent AND consequent
	supportAC = np.logical_and(antecedent, consequent).mean()

	# Compute support for antecedent
	supportA = antecedent.mean()

	# Compute support for NOT consequent
	supportnC = 1.0 - consequent.mean()

	# Compute support for antecedent and NOT consequent
	supportAnC = supportA - supportAC

    # Return conviction
	return supportA * supportnC / supportAnC
```

###**Association et dissociation**

####**3. Introduction à la métrique de Zhang**
+ ***La métrique dont nous parlerons dans cette vidéo a été introduite dans Zhang (2000) et est depuis connue sous le nom de métrique de Zhang***. 

+ **Il est borné par le bas par -1 et borné par le haut par 1. Une valeur de 1 indique une association parfaite. Le négatif 1 indique une dissociation parfaite***. 
+ ***La métrique de Zhang est complète dans le sens où elle mesure à la fois l’association et la dissociation. Il est également interprétable et a une définition en termes de métriques plus simples.**

  + Zhang, T. (2000). Règlement de l’association. Actes de la 4ème conférence Pacifique-Asie, PADKK, pp.245-256. Kyoto, Japon.


####**4. Définition de la métrique de Zhang**
+ ***La métrique de Zhang est définie comme la différence entre les métriques de confiance de « si A alors B » et « si PAS A alors B », divisée par le maximum de la confiance de « si A alors B » et de « si PAS A alors B »***. 

+ Déballons cela lentement, en commençant par le numérateur. 

+ Tout d’abord, nous calculons la métrique de confiance pour la règle « si A alors B », qui mesure le degré d’association. 

+ Et deuxièmement, nous soustrayons de cela la confiance de « si ce n’est pas A alors B », ce qui signifie la dissociation entre A et B. Les deux chiffres ne seront pas négatifs. 

+ Enfin, nous divisons par le maximum des deux mesures de confiance que nous avons calculées. Cela garantira que la valeur de la métrique n’est jamais inférieure à -1 et jamais supérieure à 1.

####**5. Construire la métrique de Zhang à l’aide du support**
+ Lorsque nous construisons la métrique de Zhang, nous utiliserons en fait une forme plus simple, qui implique exclusivement des calculs de support. 

+ Dans la version ci-dessus, le numérateur est donné comme le support de A et B moins le produit du support de A et du support de B. 

+ Le dénominateur est le maximum du support de A et B, multiplié par un moins le support de A, et le support de A multiplié par le support de B moins le support de A et B.

####**6. Calcul de la métrique de Zhang**
+ Disons que nous décidons d’utiliser la métrique de Zhang comme moyen de commander le site de livres électroniques de notre client.

####**Définition de la métrique de Zhang**

+ En général, lorsque nous voulons effectuer une tâche plusieurs fois, nous écrivons une fonction plutôt que de coder chaque instance individuelle. 

+ Dans cet exercice, nous allons définir une fonction pour la métrique de Zhang qui prend un antécédent et un conséquent et génère la métrique elle-même. 

+ Lorsque les problèmes que nous résolvons deviennent de plus en plus compliqués dans le chapitre suivant, disposer d'un moyen pratique de calculer une métrique simplifiera grandement les choses.

+ *Notez que numpy a été importé en tant que np et pandas a été importé en tant que pd. De plus, rappelez-vous que l'expression de la métrique de Zhang en termes de calculs de support est la suivante* :

$$
Zhang(A \rightarrow B) = \frac{Support(A \& B) - Support(A) Support(B)}{ max[Support(AB) (1-Support(A)), Support(A)(Support(B)-Support(AB))]}
$$
 

####**Instructions**
+ Définissez les valeurs de support de l'antécédent et du conséquent individuellement.
+ Définissez le support de {antécédent, conséquent}.
+ Complète les expressions du numérateur et du dénominateur.
+ Complétez l'expression de la métrique de Zhang.

```python
# Define a function to compute Zhang's metric
def zhang(antecedent, consequent):
	# Compute the support of each book
	supportA = antecedent.mean()
	supportC = consequent.mean()

	# Compute the support of both books
	supportAC = np.logical_and(antecedent, consequent).mean()

	# Complete the expressions for the numerator and denominator
	numerator = supportAC - supportA*supportC
	denominator = max(supportAC*(1-supportA), supportA*(supportC-supportAC))

	# Return Zhang's metric
	return numerator / denominator
```

###**L’algorithme Apriori**

####**1. L’algorithme Apriori**

+ **Dans cette vidéo, nous discuterons de l’algorithme Apriori, qui nous aidera à réduire la complexité des grands ensembles de données en éliminant les ensembles d’éléments à faible prise en charge.**

####**2. Comptage des ensembles d’objets**


+ Nous commencerons par revenir sur le problème de l’analyse du panier d’achat: 
  + il y a beaucoup d’articles et, comme nous le savons au chapitre 1, 
  + de nombreux articles impliquent de nombreuses règles d’association. 
  
+ Comme nous le verrons bientôt, l’algorithme Apriori taille les ensembles d’éléments, plutôt que les règles d’association. 

+ Il y a moins d’ensembles d’éléments que de règles, mais le nombre reste énorme. 

+ En particulier, le nombre d’ensembles d’éléments que nous pouvons générer en choisissant $k$ éléments à partir d’un ensemble de $n$ est donné par $n$ factorielle sur $n$ moins $k$ factorielle fois $k$ factorielle. 

+ C’est le nombre de « combinaisons » et est indiqué comme $n$ choisir $k$. 

+ Avec les 3461 articles de notre jeu de données sur les cadeaux de fantaisie, il y a 1 jeu d’éléments, l’ensemble nul, qui ne contient aucun élément, 3461 ensembles d’éléments avec 1 élément, près de 6 millions avec 2, près de 7 milliards avec 3 et près de 6 billions avec 4.

####**3. Comptage des ensembles d’objets**

+ Que se passe-t-il si nous autorisons un certain nombre d’articles? 

+ Nous pouvons obtenir cela en additionnant toutes les combinaisons. 

+ Le nombre total d’ensembles d’éléments de n’importe quelle taille que nous pouvons générer à partir de n éléments est de 2 au nième. 

+ Pour l’ensemble de données cadeaux, c’est 2 au 3461st. Ce nombre est beaucoup plus grand que 10 au 82ème, qui est le nombre estimé d’atomes dans l’univers observable.

####**4. Réduction du nombre d’ensembles d’éléments**

+ Si le nombre d’ensembles d’objets est supérieur au nombre d’atomes dans l’univers, il sera bien sûr impossible de tous les évaluer. 

+ En fait, nous ne pouvons même pas tous les énumérer. 

+ Mais comment pouvons-nous supprimer un ensemble d’éléments sans même l’évaluer au préalable ?

+  Nous pourrions utiliser une règle grossière, comme exclure tous les ensembles d’éléments avec plus de $k$ éléments, mais comme nous l’avons vu, ce $k$ devrait probablement être trois, ce qui éliminerait beaucoup de règles intéressantes.

+ *Notez que les itemsets fréquents ont été calculés pour vous et sont disponibles en tant que frequent_itemsets. De plus, zhangs_rule() a été défini et association_rules() a été importé de mlxtend. Vous commencerez par recalculer l'ensemble de règles d'origine. Après cela, vous appliquerez la métrique de Zhang pour sélectionner uniquement les règles avec une association élevée et positive.*

####**Instructions**

+ Calculez la règle de Zhang et affectez la sortie à la colonne zhang dans les règles.
+ Sélectionnez les règles dont la métrique de Zhang est supérieure à 0,98.

```python
import numpy as np
```

```python
# Funtion to compute Zhang's rule from mlxtend association_rules output
def zhangs_rule(rules):
    PAB = rules['support'].copy()
    PA = rules['antecedent support'].copy()
    PB = rules['consequent support'].copy()
    NUMERATOR = PAB - PA*PB
    DENOMINATOR = np.max((PAB*(1-PA).values,PA*(PB-PAB).values), axis = 0)
    return NUMERATOR / DENOMINATOR
```

```python
# Generate the initial set of rules using a minimum lift of 1.00
rules = association_rules(frequent_itemsets, metric = "lift", min_threshold = 1.00)

# Set antecedent support to 0.005
rules = rules[rules['antecedent support'] > 0.04]

# Set consequent support to 0.005
rules = rules[rules['consequent support'] > 0.04]

# Compute Zhang's rule
rules['zhang'] = zhangs_rule(rules)

# Set the lower bound for Zhang's rule to 0.98
rules = rules[rules['zhang'] > 0.5]
print(rules[['antecedents', 'consequents']])
```

####**2. Qu’est-ce qu’un tracé de coordonnées parallèles?**

+ **Alors, qu’est-ce qu’un tracé de coordonnées parallèles?** 

+ ***Nous pouvons le considérer comme une sorte de diagramme de réseau dirigé***. 

+ ***Autrement dit, le tracé montre les connexions entre deux objets qui sont liés et indique la direction de la relation***. 

+ L’exemple de tracé de coordonnées parallèles illustré a été construit à l’aide de règles d’association générées à partir du jeu de données MovieLens. 

+ Chaque ligne de l’intrigue correspond à un film particulier. 

+ Et chaque ligne relie un antécédent du côté gauche avec un postérieur du côté droit. 

+ Notez que les étiquettes sur le côté conséquent de la parcelle sont identiques à celles sur le côté antérieur. 

+ Nous pouvons voir que certains films, comme The Dark Knight, sont des antécédents pour beaucoup d’autres. 

+ D’autres films, tels que Fight Club, ne sont des antécédents que pour un seul. 

+ Les tracés de coordonnées parallèles peuvent être étendus pour incorporer plusieurs antécédents ou conséquences, mais nous nous concentrerons sur les relations à un antécédent, à un conséquent.

####**3. Quand utiliser des diagrammes de coordonnées parallèles**

+ Quand devrions-nous utiliser un diagramme de coordonnées parallèles, plutôt qu’une carte thermique ou un nuage de points? 

+ Si nous n’avons pas besoin d’informations sur l’intensité, si nous voulons seulement des informations sur l’existence d’une règle et si nous voulons un diagramme avec un encombrement visuel minimal, il est logique d’utiliser un diagramme de coordonnées parallèles, plutôt qu’une carte thermique. 

+ Alternativement, si nous voulons des informations sur des règles individuelles, que nous ne sommes pas intéressés par les seuils métriques utilisés pour identifier les règles et que nous voulons uniquement examiner les règles elles-mêmes, nous voudrons utiliser un diagramme de coordonnées parallèles, plutôt qu’un nuage de points.

De plus, apriori(), association_rules() et parallel_coordinates() ont été importés et pandas est disponible en tant que pd. La fonction rules_to_coordinates() a été définie et est disponible.*

####**Instructions**
+ Complétez l'énoncé de l'algorithme Apriori en utilisant une prise en charge minimale de 0,05.
+ Calculez les règles d'association en utilisant un seuil de confiance minimum de 0,50. 
+ Ceci est suffisamment élevé pour capturer exclusivement des points proches de la partie supérieure de la frontière de confiance de l'offre.
+ Convertissez les règles en coordonnées.
+ Tracez les coordonnées en utilisant parallel_coordinates().

```python
# Function to convert rules to coordinates.
def rules_to_coordinates(rules):
    rules['antecedent'] = rules['antecedents'].apply(lambda antecedent: list(antecedent)[0])
    rules['consequent'] = rules['consequents'].apply(lambda consequent: list(consequent)[0])
    rules['rule'] = rules.index
    return rules[['antecedent','consequent','rule']]
```

```python
# Compute the frequent itemsets
frequent_itemsets = apriori(onehot, min_support = 0.05, 
                         use_colnames = True, max_len = 2)

# Compute rules from the frequent itemsets with the confidence metric
rules = association_rules(frequent_itemsets, metric = 'confidence', 
                          min_threshold = 0.50)

# Convert rules into coordinates suitable for use in a parallel coordinates plot
coords = rules_to_coordinates(rules)

# Generate parallel coordinates plot
parallel_coordinates(coords, 'rule')
plt.legend([])
plt.show()
```
