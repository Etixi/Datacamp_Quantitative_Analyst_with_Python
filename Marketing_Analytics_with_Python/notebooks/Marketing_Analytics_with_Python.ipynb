####**4. Changer le type de données d'une colonne**

+ Pour changer le type de données d'une colonne, vous pouvez utiliser la méthode **astype()**. 

+ Par exemple, vous avez vu sur la diapositive précédente que la colonne convertie est stockée en tant qu'objet. 

+ Il contient des valeurs True et False, il est donc plus approprié de le stocker en tant que booléen. 

+ Vous pouvez utiliser la méthode **astype() avec l'argument 'bool'** comme indiqué ici pour changer son type de données. 

+ Si vous vérifiez à nouveau le type de données de la colonne 'convertie', vous verrez qu'elle est maintenant 'bool'.







```python
 #Change the data type of a column
marketing['converted'] = marketing['converted']\
                                  .astype('bool')
print(marketing['converted'].dtype)
```

####**5. Création de nouvelles colonnes booléennes**

+ La colonne marketing_channel capture le canal sur lequel un utilisateur a vu un atout marketing. 

+ Supposons que vous souhaitiez disposer d'une colonne identifiant si un élément marketing particulier était une annonce auto-promotionnelle ou non. 

+ Vous pouvez utiliser la fonction **where()** de numpy pour créer une nouvelle colonne booléenne pour l'établir. 

+ Le premier argument est une expression qui vérifie si la valeur de la colonne marketing_channel est "House Ads", le deuxième argument est la valeur que vous souhaitez attribuer si l'expression est vraie et le troisième argument est la valeur que vous souhaitez attribuer si le l'expression est fausse.

```python
import numpy as np

marketing['is_house_ads'] = np.where(    
    marketing['marketing_channel'] == 'House Ads', True, False)
print(marketing.is_house_ads.head(3))
```

####**6. Mappage des valeurs aux colonnes existantes**

+ En raison de la façon dont pandas stocke les données, dans un grand ensemble de données, il peut être inefficace en termes de calcul de stocker des colonnes de chaînes. 

+ Dans de tels cas, il peut accélérer les choses en stockant ces valeurs sous forme de nombres. Pour créer une colonne avec des codes de canaux, créez un dictionnaire qui mappe les canaux aux codes numériques. 

+ Ensuite, utilisez la méthode **map()**sur la colonne channel avec ce dictionnaire, comme illustré ici.
 


```python
channel_dict = {"House Ads": 1, 
                "Instagram": 2, 
                "Facebook": 3, 
                "Email": 4, 
                "Push": 5}
marketing['channel_code'] = marketing['marketing_channel']\
                                        .map(channel_dict)
print(marketing['channel_code'].head(3))
```

####**2. Combien d'utilisateurs voient les ressources marketing ?**

+ Pour commencer, voyons combien d'utilisateurs uniques voient nos ressources marketing chaque jour. 

+ Nous pouvons utiliser la méthode **groupby()** sur le DataFrame marketing. 

+ Pour grouper par date, nous passons 'date_servi' comme argument à groupby(). 

+ Ensuite, nous sélectionnons la colonne user_id en dehors de **groupby()** et utilisons la méthode **nunique()** pour compter le nombre d'utilisateurs uniques chaque jour.

```python
# Aggregate unique users that see ads by date

daily_users = marketing.groupby(['date_served'])\
                            ['user_id'].nunique()    

print(daily_users)
```

+ Vous pouvez ajouter un titre à l'aide de **plt dot title()** et ajouter des étiquettes **x et y** à l'aide des fonctions **plt dot xlabel() et plt dot ylabel()**, respectivement.

```python
import matplotlib.pyplot as plt
# Plot
daily_users.plot(figsize=(15,12))
# Annotate
plt.title('Daily number of users who see ads')
plt.xlabel('Date')
plt.ylabel('Number of users')
plt.xticks(rotation = 45)
plt.show()
```

####**2. La campagne a-t-elle été couronnée de succès?**

+ En tant que professionnel des données, vous vous retrouverez souvent à traduire des questions commerciales en résultats mesurables. 

+ L’une des questions les plus fréquemment posées est la suivante : **« La campagne a-t-elle été couronnée de succès ? »**. 

+ Au début, cela peut sembler une question intimidante.

+ Il y a tellement de façons de mesurer le succès de la campagne! Cela dit, il existe quelques types de mesures que vous vous retrouverez à utiliser encore et encore. Pour une équipe marketing, le succès d’une campagne est généralement mesuré par le taux de conversion. 

+ C’est-à-dire que de toutes les personnes qui sont entrées en contact avec votre campagne marketing, combien ont acheté le produit? Selon l’entreprise, cela peut signifier qu’une personne a effectué un achat ou s’est abonnée à votre service. 

+ En outre, de nombreuses entreprises d’abonnement se soucient de la **rétention**.

+ Cette mesure peut être particulièrement difficile à mesurer car elle nécessite de la patience. 

+ Nous ne pouvons pas connaître les taux de rétention de 90 jours avant que 90 jours ne se soient écoulés depuis qu’un utilisateur s’est initialement abonné.

####**3. Taux de conversion**

+ Le taux de conversion est le pourcentage de personnes que nous commercialisons et qui se convertissent finalement à notre produit. 

+ Dans ce cours, nous nous concentrerons sur un service d’abonnement et parlerons de conversion en termes d’abonnements.

$Taux de conversion  = \frac{Nombre de personnes qui se convertissent}{Nombre total de personnes auxquelles nous avons fait du marketing}$.

####**4. Calcul du taux de conversion à l’aide de pandas**

+ **Pour calculer le nombre total de personnes qui ont converti, nous découpons le DataFrame pour inclure uniquement les lignes où « converti » est égal à True, puis comptons les user_ids uniques à l’aide de nunique()**. 
+ Ensuite, nous calculons le nombre total de personnes auxquelles nous avons commercialisé. 

+ Nous pouvons le faire en comptant tous les ID utilisateur uniques dans notre jeu de données. 

+ Maintenant que nous avons le nombre total d’utilisateurs et d’utilisateurs qui se sont abonnés, nous pouvons calculer le taux de conversion en divisant les abonnés par le total.



```python
# Import pandas into the environment
import pandas as pd

# Import marketing.csv 
marketing = pd.read_csv('/content/marketing.csv')
marketing.head(3)
```

```python
subscribers = marketing[marketing['converted'] == True]\
                                  ['user_id'].nunique()
total = marketing['user_id'].nunique()
conv_rate = subscribers/total
print(round(conv_rate*100, 2), '%')
```

####**5. Taux de rétention**
+ Le taux de rétention est le pourcentage de personnes qui restent abonnées après un certain temps. 

+ Dans ce cours, nous nous concentrerons sur la rétention de 1 mois.

$Taux de rétention = \frac{Nombre de personnes qui restent abonnées}{Nombre total de personnes qui se sont converties}$.
####**6. Calcul du taux de rétention**

+ **Pour déterminer le taux de rétention, nous calculons d’abord le nombre total d’utilisateurs qui restent abonnés après un mois.** 

+ Cette information est disponible dans la colonne **« is_retained »**. 

+ Nous découpons maintenant le DataFrame pour inclure uniquement les lignes où l’utilisateur a été conservé, c’est-à-dire où **is_retained est égal à True et compter les user_ids uniques**. 

+ Nous pouvons réutiliser notre calcul « abonnés » à partir du calcul du taux de conversion, car le nombre total d’utilisateurs qui se sont abonnés à l’origine reste le même. 

+ Enfin, nous divisons le nombre d’utilisateurs qui ont été retenus par le nombre d’utilisateurs qui se sont abonnés pour calculer le taux de rétention.


```python
retained = marketing[marketing['is_retained'] == True]\
                                ['user_id'].nunique()
subscribers = marketing[marketing['converted'] == True]\
                                ['user_id'].nunique()
retention = retained/subscribers
print(round(retention*100, 2), '%')
```

```python
# Count the subs by referring channel and day
retention_total = marketing.groupby(['date_subscribed','subscribing_channel'])\
                                    ['user_id'].nunique()
                                    
retention_subs = marketing[marketing['is_retained'] == True]\
                .groupby(['date_subscribed','subscribing_channel'])\
                ['user_id'].nunique()

# Calculate the retention rate
daily_retention_rate = retention_subs/retention_total

daily_retention_rate = pd.DataFrame(daily_retention_rate.unstack(level=1))
daily_retention_rate
```

```python
def retention_rate(dataframe, column_names):
  # Group by column_names and calculate retention    
  retained = dataframe[dataframe['is_retained'] == True]\
  .groupby(column_names)['user_id'].nunique()
  # Group by column_names and calculate conversion    
  converted = dataframe[dataframe['converted'] == True]\
  .groupby(column_names)['user_id'].nunique()    
  retention_rate = retained/converted
  return retention_rate
```

```python
#Retention rate by channel 
daily_retention = retention_rate(marketing,
                                 ['date_subscribed', 'subscribing_channel'])
daily_retention = pd.DataFrame(daily_retention.unstack(level=1))
print(daily_retention.head())
```

```python
def conversion_rate(dataframe, column_names):
    # Total number of converted users
    column_conv = dataframe[dataframe['converted'] == True]\
                       .groupby(column_names)['user_id'].nunique()

    # Total number users
    column_total = dataframe.groupby(column_names)['user_id'].nunique()   
    
    # Conversion rate 
    conversion_rate = column_conv/column_total
    
    # Fill missing values with 0
    conversion_rate = conversion_rate.fillna(0)
    return conversion_rate
```

+ Le lift peut être calculé en calculant la différence entre l'effet de traitement (ou la moyenne) du traitement par rapport à l'effet de traitement du contrôle divisé par l'effet de traitement du contrôle. 

+ La formule de l'ascenseur peut être trouvée ci-dessous:

 $$
  \frac{\text{Treatment conversion rate - Control conversion rate}}{\text{Control conversion rate}}
 $$

+ Le résultat est la différence en pourcentage entre le contrôle et le traitement.

```python
def lift(a,b):
    # Calcuate the mean of a and b
    a_mean = np.mean(a)
    b_mean = np.mean(b)
    
    # Calculate the lift using a_mean and b_mean
    lift = (b_mean  - a_mean)/ a_mean
  
    return str(round(lift*100, 2)) + '%'
  
# Print lift() with control and personalization as inputs
print(lift(control, personalization))
```

```python
def ab_segmentation(segment):
  # Build a for loop for each subsegment in marketing
  for subsegment in np.unique(marketing[segment]):
      print(subsegment)
      
      # Limit marketing to email and subsegment
      email = marketing[(marketing['marketing_channel'] == 'Email') & (marketing[segment] == subsegment)]

      subscribers = email.groupby(['user_id', 'variant'])['converted'].max()
      subscribers = pd.DataFrame(subscribers.unstack(level=1)) 
      control = subscribers['control'].dropna()
      personalization = subscribers['personalization'].dropna()

      print('lift:', lift(control, personalization)) 
      print('t-statistic:', stats.ttest_ind(control, personalization), '\n\n')
```
