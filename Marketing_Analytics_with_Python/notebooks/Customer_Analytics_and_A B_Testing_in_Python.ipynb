####**4. Données de regroupement : .groupby()**

+ Nous pouvons appeler la méthode 'groupby()' sur un dataFrame pour spécifier les groupes à agréger. 

+ Ici, nous l’utiliserons sur notre ensemble de données démographiques et d’achat combinées. 

+ L’argument principal est 'by' auquel nous fournissons une liste de champs dataFrame sur lesquels nous voulons regrouper. 

+ Ici, les champs potentiellement pertinents sont « pays », « appareil », « sexe » et « âge ». 

+ Regroupons par « pays » et « appareil ». L’argument pertinent suivant est 'axis' qui spécifie si nous regroupons par valeurs de ligne ou de colonne. La valeur par défaut, « 0 », regroupe par colonnes, ce que nous allons faire ici, et pour le reste du cours. 

+ L’autre argument intéressant est « as_index ». Par défaut, cet argument est « True », ce qui signifie que les champs regroupés par deviennent des index. 

+ Nous voulons définir cela sur « Faux » afin que cela ne se produise pas. Cela renvoie un objet dataFrameGroupedBy. L’étape suivante consiste à agréger ces groupes.

####**5. Agrégation des données - prix moyen payé par groupe**

+ Pour ce faire, la méthode la plus simple consiste à appeler une méthode d’agrégation sur l’objet dataFrameGroupedBy. 

+ Appelons 'moyenne' sur la valeur 'price' de notre dataFrame. La sortie est le montant moyen payé par abonnement pour tous les utilisateurs acheteurs. 

+ Dans ce cas, plutôt que d’être calculé sur l’ensemble des données, il est calculé sur chaque combinaison de pays d’appareil.

####**6. Données agrégées : .agg()**

+ Toute fonction intégrée similaire à mean peut être appelée sur un objet dataFrameGroupedBy. 

+ Cependant, des options plus flexibles existent via la méthode point agg() La façon la plus simple d’utiliser cette méthode est de lui passer une fonction comme moyenne. 

+ Comme nous pouvons le voir, cela a le même résultat que lorsque nous appelions signifie directement.

####**7. .agg(): fonctions multiples**

+ Il peut être encore élargi en passant une liste de fonctions, comme la moyenne et la médiane, et en calculant les deux.

####**8. .agg(): plusieurs fonctions, plusieurs colonnes**

+ La vraie flexibilité vient d’un troisième type d’argument. 

+ Nous pouvons passer dans un dictionnaire où les clés sont des noms de colonne dans notre ensemble de données, comme « achat » ou « âge » et les valeurs sont une liste de fonctions à appliquer sur ces colonnes; encore répartis par groupes. 

+ Trouvons la moyenne, la valeur minimale et maximale de chacun des achats et de l’âge, à titre d’exemple.

####**9. .agg(): fonctions personnalisées**

+ Une autre grande flexibilité de la méthode agg est que nous pouvons également transmettre nos propres fonctions pour les agréger, pas seulement celles intégrées. 

+ Voici une fonction qui trouve la valeur moyenne tronquée, c’est-à-dire qu’elle supprime les dix pour cent supérieurs et inférieurs des valeurs avant de calculer la moyenne. 

+ Nous pouvons agréger notre âge sur le pays et les groupes d’appareils avec cette fonction. 

+ La seule distinction est que lors de la saisie de cette fonction, nous ne voulons pas que son nom entre guillemets comme nous l’avons fait pour les fonctions intégrées.

```python
def get_pvalue(con_conv, test_conv, con_size, test_size):  
    lift =  - abs(test_conv - con_conv)

    scale_one = con_conv * (1 - con_conv) * (1 / con_size)
    scale_two = test_conv * (1 - test_conv) * (1 / test_size)
    scale_val = (scale_one + scale_two)**0.5

    p_value = 2 * norm.cdf(lift, loc = 0, scale = scale_val )

    return p_value
```

```python
def get_ci(value, cl, sd):

    loc = norm.ppf(1 - cl/2)
    rng_val = norm.cdf(loc - value/sd)

    lwr_bnd = value - rng_val
    upr_bnd = value + rng_val 

    return_val = (lwr_bnd, upr_bnd)
    return(return_val)
```
