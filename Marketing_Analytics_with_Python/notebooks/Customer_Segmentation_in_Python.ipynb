####**4. Qu’est-ce que l’analyse de cohorte?**

+ **L’analyse de cohorte est un outil d’analyse descriptive. Il regroupe les clients en cohortes mutuellement exclusives - qui sont ensuite mesurées au fil du temps. L’analyse de cohorte fournit des informations plus approfondies que les mesures dites de vanité. Il aide à mieux comprendre les tendances de haut niveau en fournissant des informations sur les métriques du cycle de vie du produit et du client.**

####**5. Types de cohortes**

+ **Il existe des types de cohortes arborescentes**. 

+ **Les cohortes de temps sont des clients qui se sont inscrits à un produit ou à un service au cours d’une période donnée**. 
  + **L’analyse de ces cohortes montre le comportement des clients en fonction du moment où ils ont commencé à utiliser les produits ou services de l’entreprise. L’heure peut être mensuelle ou trimestrielle, voire quotidienne**. 
  
+ **Les cohortes comportementales sont des clients qui ont acheté un produit ou souscrit à un service dans le passé.** 
  ***Il regroupe les clients en fonction du type de produit ou de service qu’ils ont souscrit.*** 

  + ***Les clients qui se sont inscrits à des services de niveau de base peuvent avoir des besoins différents de ceux qui se sont inscrits à des services avancés.*** 

+ **Comprendre les besoins des différentes cohortes peut aider une entreprise à concevoir des services ou des produits sur mesure pour des segments particuliers.** 

+ **Les cohortes de taille font référence aux différentes tailles de clients qui achètent les produits ou services de l’entreprise**. 
  + ***Cette catégorisation peut être basée sur le montant des dépenses dans une certaine période de temps après l’acquisition, ou le type de produit que le client a dépensé la majeure partie du montant de sa commande dans une certaine période de temps.***

```python
def get_month(date):
    return dt.datetime(date.year, date.month, 1)

online['InvoiceMonth'] = pd.to_datetime(online['InvoiceDate']).apply(get_month)
grouping = online.groupby('CustomerID')['InvoiceMonth']
online['CohortMonth'] = grouping.transform('min')
```

```python
# Define a function that will parse the date
def get_day(x): return dt.datetime(x.year, x.month, x.day) 

# Create InvoiceDay column
online['InvoiceDay'] = online['InvoiceDate'].apply(get_day) 

# Group by CustomerID and select the InvoiceDay value
grouping = online.groupby('CustomerID')['InvoiceDay'] 

# Assign a minimum InvoiceDay value to the dataset
online['CohortDay'] = grouping.transform('min')

# View the top 5 rows
display(online.head())
```

####**Calculer le décalage horaire en jours - partie 2**

```python
# Import StandardScaler 
from sklearn.preprocessing import StandardScaler

# Apply log transformation
datamart_rfmt_log = np.log(datamart_rfmt)

# Initialize StandardScaler and fit it 
scaler = StandardScaler(); scaler.fit(datamart_rfmt_log)

# Transform and store the scaled data as datamart_rfmt_normalized
datamart_rfmt_normalized = scaler.transform(datamart_rfmt)
```

####**Calculer et tracer la somme des carrés des erreurs**

+ ***Maintenant, vous allez calculer la somme des erreurs au carré pour différents nombres de clusters allant de 1 à 10.***

+ **Vous utiliserez les données RFMT normalisées que vous avez créées dans l'exercice précédent, elles sont stockées sous datamart_rfmt_normalized. Le module KMeans de scikit-learn est également importé. De plus, nous avons initialisé un dictionnaire vide pour stocker la somme des erreurs au carré sous la forme sse = {}.**

+ *N'hésitez pas à explorer la date dans la console.*

####**Instructions**
+ Initialisez KMeans avec $k$ clusters et l'état aléatoire 1 et ajustez KMeans sur l'ensemble de données normalisé.
+ Attribuez la somme des distances au carré à l'élément k du dictionnaire sse.
+ Ajoutez le titre du tracé "La méthode du coude", l'étiquette de l'axe $X$ "k" et l'étiquette de l'axe $Y$ "SSE".
+ Tracez les valeurs SSE pour chaque $k$ stocké sous forme de clés dans le dictionnaire.

```python
# Fit KMeans and calculate SSE for each k between 1 and 10
sse={}
for k in range(1, 11):
  
    # Initialize KMeans with k clusters and fit it 
    kmeans = KMeans(n_clusters=k, random_state=1 ).fit(datamart_rfmt_normalized)
    
    # Assign sum of squared distances to k element of the sse dictionary
    sse[k] = kmeans.inertia_  

# Add the plot title, x and y axis labels
plt.figure(figsize=(18,15))
plt.title('The Elbow Method')
plt.xlabel('k')
plt.ylabel('SSE')

# Plot SSE values for each k stored as keys in the dictionary
sns.pointplot(x=list(sse.keys()), y=list(sse.values()))
plt.show()
```
