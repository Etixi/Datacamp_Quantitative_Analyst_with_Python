####**3. Parties et étapes des données d’apprentissage supervisé**

+ Les modèles d’apprentissage supervisé nécessitent deux éléments de données clés.

+ ***Il pourrait s’agir de prédire quels clients vont se désabonner ou quels clients achèteront à nouveau.***

####**4. Format des données sur l’apprentissage supervisé**

+ Le format de données pour le modèle d’apprentissage supervisé est généralement une matrice $X$ par $N$ plus 1 avec $X$ nombre d’observations

####**2. Introduction à la régression logistique**

+ **La régression logistique est une technique d’apprentissage supervisé qui prédit les variables de réponse binaires. La régression logistique modélise le logarithme du rapport de cotes.** 

+ **Les cotes sont un rapport de la probabilité que l’événement se produise divisé par la probabilité que l’événement ne se produise pas, ou $p$ divisé par $1 moins p$, soit $\frac{p}{1-p}$**. 

+ Par exemple, si la probabilité de désabonnement est de 75%, la probabilité de ne pas attrier est de 25%, donc le rapport de cotes sera de 75% divisé par 25% ou 3. Le logarithme de 3 est d’environ 0,48. 

+ Les raisons derrière les mathématiques dépassent la portée de ce cours, mais cette approche aide à trouver la limite de décision entre les deux classes tout en gardant les coefficients linéairement liés à la variable cible. 

+ Voici la formule de l’équation de régression logistique basée sur deux variables d’entrée et une probabilité $p$.

####**Régression logistique**
+ Modèle de classification pour une variable cible binaire
+ Modélise le logarithme du rapport de cotes
  + Rapport de cotes : La probabilité qu’un evenement se produise divisée par la probabilité que l’événement ne se produise pas  $\frac{p}{1−p}$
 
  + Par exemple:

    + Si la probabilité de désabonnement est  75%
    + alors la probabilité de ne pas se désabonner est  25%
    + par conséquent, le rapport de cotes sera  75% divisé par  25% ou  3
  
      + La base  10
      + logarithme de  3
      + $log_{10}(3)$ est à peu près  0.48 .

+ Cette approche permet de trouver la limite de décision entre les 2 classes tout en gardant les coefficients linéairement liés à la variable cible.

```python
prob_churn = 0.75
prob_no_churn = 1 - 0.75

odds = prob_churn / prob_no_churn 
log_odds = np.log10(odds).round(2)

print(f'Probability of churn = {prob_churn} \nProbability of no churn = {prob_no_churn} \nOdds ratio = {odds} \nLog odds = {log_odds}')
```

```python
from sklearn.metrics import recall_score, precision_score

C = [1, 0.5, 0.25, 0.1, 0.05, 0.025, 0.01, 0.005, 0.0025]
l1_metrics = np.empty((len(C),3), float)
```

```python
# Run a for loop over the range of C list length
for index in range(0, len(C)):
  # Initialize and fit Logistic Regression with the C candidate
  logreg = LogisticRegression(penalty='l1', C=C[index], solver='liblinear')
  logreg.fit(train_X, train_Y)
  # Predict churn on the testing data
  pred_test_Y = logreg.predict(test_X)
  # Create non-zero count and recall score columns
  l1_metrics[index,1] = np.count_nonzero(logreg.coef_)
  l1_metrics[index,2] = recall_score(test_Y, pred_test_Y)

# Name the columns and print the array as pandas DataFrame
col_names = ['C','Non-Zero Coeffs','Recall']
print(pd.DataFrame(l1_metrics, columns=col_names))
```

###**Principes de base de la valeur vie client (CLV)**

####**1. Principes de base de la valeur vie client (CLV)**

+ ***Excellent travail sur la prédiction de l’attrition des clients! Dans ce chapitre, nous explorerons le concept de valeur vie client et différentes façons de le calculer. Ensuite, nous apprendrons à utiliser la régression pour prédire les transactions des clients au cours du mois prochain.***

####**2. Qu’est-ce que la CLV ?**

+ **La valeur à vie du client est une mesure de ce qu’une entreprise s’attend à gagner d’un client moyen au cours de sa vie.*** 
+ **Il peut être historique - où nous résumons le profit de chaque client et obtenons la valeur réelle du client**. 

+ **Une autre alternative consiste à prédire la valeur à vie du client, ce qui est un peu plus complexe.** 

+ **Il existe un certain nombre de méthodes pour prédire la valeur à vie du client**. 

+ **Nous ne nous pencherons pas sur les méthodes de prédiction complexes telles que les modèles de Pareto / NBD ou les modèles bêta-géométriques / NBD qui prendraient un cours complet d’elles-mêmes***. 

+ **Au lieu de cela, nous utiliserons des méthodes traditionnelles simples pour obtenir des estimations décentes. En outre, nous apprendrons à prédire des variables continues comme le nombre d’achats avec régression.**

####**3. CLV historique**

+ ***La valeur historique à vie du client est simplement la somme des revenus des transactions des clients, multipliée par la marge bénéficiaire moyenne ou au niveau du produit***. 

+ ***Alternativement, nous pouvons additionner le bénéfice réel de chaque achat, si ces données sont disponibles.*** 

+ ***Le principal défi de l’approche historique est qu’elle ne tient pas compte de la tenure, de la fidélisation et des taux de désabonnement des clients***. 

+ Si l’entreprise acquiert beaucoup de nouveaux clients, leur valeur historique moyenne à vie sera déflatée en raison de leur courte durée. 

+ **Le deuxième défi est que les nouveaux clients sont traités de la même manière que les clients titulaires, de sorte que la valeur moyenne à vie du client ne représente pas le revenu futur potentiel.**

####**4. Formule CLV de base**
+ Le calcul de la valeur à vie le plus élémentaire consiste à prendre le revenu moyen par client au cours d’une certaine période, disons un mois, à le multiplier par la marge bénéficiaire, puis à multiplier le résultat par la durée de vie moyenne ou attendue du client. 

+ ***La durée de vie peut être définie à l’aide des connaissances de l’entreprise sur ses clients ou en analysant le temps moyen nécessaire aux clients pour se désabonner.***

####**5. Formule CLV granulaire**
+ Une version plus granulaire de la formule de base de la valeur à vie du client examine chaque transaction. 

+ ***Ici, nous multiplions le revenu moyen par achat ou transaction par la fréquence moyenne au cours d’une période définie***
  - **par exemple un mois** 
  - **puis multiplions cela par la marge bénéficiaire.** 
  
+ Ensuite, nous multiplions le résultat par la durée de vie moyenne du client. 

+ ***Cette méthode tient compte de la fréquence des transactions dans un certain délai, ainsi que du revenu moyen par transaction, capturant ainsi des points de données plus granulaires***. 

+ ***Néanmoins, il ne tient pas compte des taux de fidélisation de la clientèle et suppose que la fréquence et le revenu par transaction resteront les mêmes au cours de la durée de vie définie.***

####**6. Formule CLV traditionnelle**


+ La formule traditionnelle est la technique descriptive de la valeur à vie client la plus populaire. 

+ Il intègre les taux de rétention et de désabonnement. 

+ Nous le calculons en multipliant le revenu moyen par la marge bénéficiaire, puis par le ratio rétention / désabonnement. 

+ Le taux de désabonnement est défini comme 1 moins la rétention. 

+ Le ratio rétention / désabonnement nous donne un multiplicateur, qui agit comme une approximation de la durée de vie prévue du client avec l’entreprise. 

+ C’est utile, mais cela suppose que le taux de désabonnement est définitif, par conséquent, le délai utilisé pour la rétention et le désabonnement est critique. 

+ En particulier, dans le cadre commercial non contractuel, nous devons nous assurer que les clients qui sont définis comme barattés dans ce délai ne reviennent pas plus tard.

```python
def get_date_int(df,col):
    year  = df[col].map(lambda x: x.year)
    month = df[col].map(lambda x: x.month)
    day   = df[col].map(lambda x: x.day)
    return year, month, day
```

```python
# Exclude target variable
online_X = online[online['InvoiceMonth']!='2011-11']

# Define the snapshot date
NOW = dt.datetime(2011,11,1)

# Calculate recency by subtracting current date from the latest InvoiceDate
features = online_X.groupby('CustomerID').agg({
    
                            # Calculate recency by subtracting the InvoiceDate from the snapshot date
                            'InvoiceDate': lambda x: (NOW - x.max()).days,
    
                            # Calculate frequency by counting unique number of invoices
                            'InvoiceNo': pd.Series.nunique,
    
                            # Calculate monetary value by summing all spend values
                            'TotalSum': np.sum,
    
                            # Calculate average and total quantity
                            'Quantity': ['mean', 'sum']}).reset_index()

# Rename the columns
features.columns = ['CustomerID', 'recency', 'frequency', 'monetary', 'quantity_avg', 'quantity_total']

features.head()
```

####**4. Mesures de performance de régression**

+ Pour mesurer la performance de régression, nous utiliserons des mesures différentes de la classification, étant donné que le résultat est une variable continue. 

+ **Plus ces mesures sont basses, meilleures sont les performances du modèle**. 

+ Les indicateurs clés pour mesurer les performances de régression sont les suivants : **Erreur quadratique moyenne racine ou RMSE en abrégé**. 

+ Il s’agit d’une racine carrée de la différence quadratique moyenne entre les valeurs prévues et réelles.

+  Nous le calculons en construisant le modèle, puis en soustrayant les valeurs prédites et réelles, en les quadraturant, en calculant la moyenne et en prenant la racine carrée pour obtenir une mesure d’erreur normalisée. 

+ La quadrature des différences est importante pour s’assurer que les erreurs positives et négatives ne s’annulent pas lors de leur répartition. 

+ **Une autre mesure est l’erreur absolue moyenne**. 

+ Il est similaire à l’erreur quadratique moyenne racine, sauf qu’ici nous prenons la moyenne des différences absolues entre les valeurs prédites et réelles. 

+ De cette façon, nous n’avons pas besoin de résoudre la quadrature des différences et la métrique est moins vulnérable aux valeurs aberrantes. 

+ **La dernière mesure couramment utilisée est le pourcentage d’erreur absolu moyen - il mesure la différence en pourcentage entre les valeurs prévues et réelles***. 

+ C’est une mesure intuitive car elle est normalisée pour être comprise entre 0 et 100%, mais elle nécessite que les valeurs cibles réelles soient supérieures à zéro pour ne pas entrer dans la division par zéro erreur.

+  Ce n’est pas le cas dans notre exemple, nous ne l’utiliserons donc pas ici.

####**5. Régression supplémentaire et mesures d’apprentissage supervisé**

+ Il existe d’autres mesures de régression que nous allons explorer. 

+ R-carré est une mesure statistique qui calcule la proportion en pourcentage de la variance inhérente qui est expliquée par le modèle. 

+ Il ne peut être calculé que pour les modèles de régression, pas pour la classification. 

+ Plus le pourcentage est élevé, mieux notre modèle explique la variance. Une autre mesure importante est la valeur p qui mesure la signification statistique de l’hypothèse dite nulle qui stipule que le modèle ou le coefficient ont été observés en raison du hasard. 

+ Plus la valeur de p est basse, mieux c’est. 

+ En règle générale, nous recherchons moins de 5 ou 10% des valeurs p pour conclure que le modèle ou les coefficients sont statistiquement significatifs.
