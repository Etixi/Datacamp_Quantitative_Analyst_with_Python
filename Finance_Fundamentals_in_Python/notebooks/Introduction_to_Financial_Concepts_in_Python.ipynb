####**3. Calcul du rendement du capital investi (% de gain)**

+ Pour commencer, commençons par une formule simple pour calculer le rendement, ou le gain sur un investissement.

####**4. Exemple**

+ Imaginez que vous avez fait un investissement de 10 000 dollars l’année 1 et que votre investissement de l’année 2 vaut 11 000 dollars. 

+ Le rendement de votre investissement est facile à calculer : 
  + soustrayez simplement vos 10 000 dollars initiaux de vos 11 000 dollars finaux pour calculer votre gain de 1 000 dollars, 
  + puis divisez ce gain par votre investissement initial pour obtenir votre taux de rendement de placement de 10 % sur votre investissement initial au cours de l’année.

####**5. Calcul du rendement du capital investi (valeur monétaire)**

+ Vous pouvez également réorganiser la formule pour calculer la valeur monétaire d’un placement en fonction d’un taux de rendement, $r$.

####**6. Exemple**
+ Par exemple, dans ce cas, si vous savez que le taux de rendement annuel de votre investissement était de 10%, vous pouvez multiplier la valeur de votre investissement, 10 000 dollars, par 1 + 10% pour obtenir 11 000 dollars, ce qui correspond aux chiffres donnés dans l’exemple précédent.

####**7. Croissance cumulative (ou amortissement)**

+ Mais que se passe-t-il si vous avez un investissement qui génère constamment un rendement année après année, croissant à chaque période qui passe? 

+ La valeur d’un investissement avec une croissance cumulative constante au fil du temps peut être calculée en fonction de $r$, le taux de rendement attendu de l’investissement (ou taux de croissance), et $t$, la durée de vie de l’investissement, et $v$, qui est la valeur initiale de l’investissement au temps 0. 

+ En utilisant ces paramètres simples, nous pouvons dériver une formule de base pour calculer la valeur de votre investissement avec un taux de croissance constant dans le temps: 
  + Valeur d’investissement = votre investissement initial multiplié par $1 + r$, à la puissance de $t$. 
  
  + Cependant, les taux de croissance sont souvent imprévisibles en finance plutôt que constants, et peuvent même être négatifs, conduisant à une dépréciation, ce qui signifie simplement une valeur décroissante au fil du temps.

####**8. Facteurs d’actualisation**

+ Mais que se passe-t-il si vous connaissez la valeur amortie d’un investissement et le taux d’amortissement, mais que vous souhaitez calculer la valeur initiale de l’investissement? 

+ En utilisant l’algèbre simple, vous pouvez construire une formule pour dériver le facteur d’actualisation (df), qui est le nombre qui, multiplié par la valeur future, est égal à la valeur initiale de l’investissement.

####**9. Intérêts composés**

+ Les périodes de capitalisation peuvent également modifier le taux de croissance d’un investissement. 

+ Par exemple, les intérêts sur votre facture de carte de crédit sont composés mensuellement. 

+ Si elle était composée quotidiennement, les cartes de crédit seraient beaucoup plus dangereuses. 

+ La formule permettant de calculer le rendement cumulatif d’un investissement avec des périodes de composition est une simple extension de la formule de l’exercice précédent, divisant r par le nombre de périodes de composition, c, et multipliant t par c également.

####**Instructions**
+ Calculez la valeur future (rendement cumulé) d'un investissement de 100 $ qui croît à un taux de 6 % par an pendant 30 années consécutives et attribuez-la à future_value.

####**Intérêts composés**

+ Comme vous l'avez vu dans l'exercice précédent, le temps et le taux de rendement sont des variables très importantes lors de la prévision de la valeur future d'un investissement.

+ Une autre variable importante est le nombre de périodes de composition, qui peut grandement affecter les rendements composés au fil du temps.

####**Instructions**
+ Calculez la valeur d'un investissement de 100 $ qui croît à un taux de 6 % par an pendant 30 années consécutives composé une fois par an et attribuez-le à l'investissement_1.
+ Calculez la valeur du même investissement, mais composée trimestriellement et attribuez-la à investissement_2.
+ Calculez la valeur du même investissement, mais composée mensuellement et attribuez-la à investissement_3.

```python
# Predefined variables
initial_investment = 100
growth_periods = 30
growth_rate = 0.06

# Calculate the value for the investment compounded once per year
compound_periods_1 = 1
investment_1 = initial_investment*(1 + growth_rate / compound_periods_1)**(compound_periods_1*growth_periods)
print("Investment 1: " + str(round(investment_1, 2)))

# Calculate the value for the investment compounded quarterly
compound_periods_2 = 4
investment_2 = initial_investment*(1 + growth_rate / compound_periods_2)**(compound_periods_2*growth_periods)
print("Investment 2: " + str(round(investment_2, 2)))

# Calculate the value for the investment compounded monthly
compound_periods_3 = 12
investment_3 = initial_investment*(1 + growth_rate / compound_periods_3)**(compound_periods_3*growth_periods)
print("Investment 3: " + str(round(investment_3, 2)))
```

####**Facteurs d'actualisation et amortissement**

+ Malheureusement, tout ne prend pas de valeur avec le temps.

+ En fait, de nombreux actifs se déprécient ou perdent de la valeur avec le temps. Pour simuler cela, vous pouvez simplement supposer un taux de rendement attendu négatif.

+ Exemple:

  + *Calculez la valeur dépréciée finale d'une voiture initialement de 10 000 $ dont la valeur diminue de 3 % par an pendant 10 ans* :

  $$
   \$10,000*(1 + -0.03)^{10} = \$7,374.24 
  $$


####**Instructions**
+ Calculez la valeur future d'un investissement de 100 $ dont la valeur se déprécie de 5 % par an pendant 10 ans et attribuez-la à future_value.

+ Calculez le facteur d'actualisation de l'investissement et affectez-le à discount_factor.

+ Dérivez la valeur initiale de l'investissement à l'aide de la valeur future et du facteur d'actualisation et affectez-la à initial_investment_again.

```python
# Calculate the future value
initial_investment = 100
growth_rate = -0.05
growth_periods = 10
```

```python
future_value = initial_investment*(1 + growth_rate)**(growth_periods)
print("Future value: " + str(round(future_value, 2)))
```

```python
# Calculate the discount factor
discount_factor = 1/((1 + growth_rate)**(growth_periods))
print("Discount factor: " + str(round(discount_factor, 2)))
```

```python
# Derive the initial value of the investment
initial_investment_again = future_value * discount_factor
print("Initial value: " + str(round(initial_investment_again, 2)))
```

####**2. La valeur non statique de l’argent**

+ La valeur de l’argent elle-même change avec le temps, et je peux le prouver avec quelques déclarations simples. 

+ Préférez-vous avoir 100 dollars dans votre poche maintenant, ou 100 dollars dans votre poche demain? 

+ Bien sûr, vous préférez avoir 100 dollars dans votre poche aujourd’hui parce qu’il est plus précieux de l’avoir dans votre main et de pouvoir en faire quelque chose d’utile aujourd’hui plutôt que demain. 

+ Maintenant, et si je vous offrais 10 000 dollars maintenant, ou 10 500 dollars dans un an ? 

+ La décision devient beaucoup plus difficile à prendre. 

+ Combien valent vraiment pour vous les 500 dollars supplémentaires gagnés en attendant un an? 

+ La raison pour laquelle il est difficile de prendre une décision est que vous comparez en fait deux types de valeurs différents, même si les deux offres sont en dollars. 

+ Les 10 000 sont en dollars de la valeur actualisée et les 10 500 en dollars futurs. Vous ne pouvez pas les comparer directement.

####**3. Le temps, c’est de l’argent**

+ Mais vous pouvez cependant estimer la valeur future en dollars des 10 000 dollars en valeur actuelle en déterminant ce que vous pourriez en faire au cours d’une année. 

+ Vous pouvez, bien sûr, prendre les 10 000 dollars et les cacher dans un compte d’épargne à votre banque, gagnant 1% par an, sans risque. 

+ C’est bien, mais vous pouvez gagner plus si vous prenez plus de risques. 

+ Prenez les 10 000 dollars, investissez-les en bourse et gagnez 8% par an en moyenne. 

+ Vous pourriez également perdre de l’argent ou gagner plus de 8%, mais en moyenne, le rendement sera d’environ 8%. Ou enfin, vous pourriez simplement prendre les 10 500 dollars si vous attendez un an.

####**4. Comparaison des valeurs futures**

+ Ce que vous avez fait en examinant ces options est calculé la valeur future de chaque scénario. 

+ Cela vous permet ensuite de comparer les scénarios puisqu’ils sont tous libellés dans la même devise temporelle, que nous appellerons dollars futurs dans cet exemple. 

+ Mettre 10 000 dollars présents à la banque vous permettra de retirer 10 100 dollars futurs dans un an. 

+ L’option B est le meilleur scénario, résultant en 10 800 dollars futurs en moyenne après 1 an en investissant 10 000 dollars présents. 

+ Cependant, il y a bien sûr un risque que votre investissement puisse valoir plus ou moins que cette valeur, mais en moyenne, nous pouvons quantifier le paiement. 

+ Cela vous laisse avec une décision basée uniquement sur le risque. 

+ Si vous ne voulez aucun risque, prenez l’option C, qui vous rapporte 10 500 dollars futurs, garantis. Si vous n’avez pas peur du risque, prenez l’option B pour un rendement moyen de 10 800 dollars futurs.

####**2. Qu’est-ce que WACC?**

+ **Vous utiliserez le WACC pour actualiser les valeurs, tout comme vous avez utilisé le taux d’inflation dans les exercices précédents**. 

+ **Le CMPC est le coût de financement d’une entreprise et comprend deux parties** : 
  + **le coût des capitaux propres et le coût de la dette**. 
  
+ **Chaque coût est multiplié par la proportion relative du financement d’une entreprise qui provient de la dette ou des capitaux propres.** 

+ **Mathématiquement parlant, le coût moyen pondéré du capital pour une entreprise donnée est égal à la somme pondérée du coût des capitaux propres et du coût de la dette de l’entreprise, y compris les implications fiscales du financement par emprunt**. 

+ **Le coût des capitaux propres est le taux de rendement requis que les investisseurs attendent pour une entreprise donnée, et le coût de la dette est le taux d’intérêt moyen pondéré pour l’encours de la dette de l’entreprise.**

####**3. Part du financement**

+ Les pondérations proportionnelles de financement de la dette et des capitaux propres peuvent être calculées à l’aide de ces formules. 

+ Pour calculer la pondération du financement par actions, vous pouvez diviser la valeur marchande des capitaux propres d’une entreprise par la valeur marchande totale d’une entreprise. 

+ Pour la pondération du financement par emprunt, vous divisez la valeur marchande du financement par emprunt d’une entreprise par la valeur marchande totale de l’entreprise.

####**5. Remises à l’aide du WACC**

+ ***Le WACC est normalement beaucoup plus élevé que l’inflation, ce qui signifie que la plupart des entreprises rechercheront un retour sur investissement supérieur à 2-3%, ce qui est le taux d’inflation typique***.

####**4. Opérations cumulées dans NumPy**

+ Dans la prochaine série d’exercices, vous utiliserez les fonctions numpy cumulées .cumsum et .cumprod. . Cumsum renvoie la somme cumulée d’un tableau de nombres, au lieu d’un seul nombre en tant que somme. 

+ Par exemple, pour le tableau 1, 2, 3, la fonction .cumsum retourne en fait un tableau de 1, 1+2 et 1+2+3, soit 1, 3 et 6. . 

+ Cumprod fonctionne de la même manière, sauf qu’au lieu d’une somme, il retourne le produit de tous les nombres du tableau. 

+ Ainsi, le même tableau 1, 2, 3 retourne maintenant 1, 1*2, 1*2*3, soit 1, 2, 6.

####**5. Prévision de la croissance cumulative**

+ Par exemple, pour calculer la valeur cumulée à chaque point dans le temps d’un investissement de 100 dollars qui augmente de 3 % dans la période 1, puis de 3 % à nouveau dans la période 2, puis de 5 % dans la période 3 - vous pouvez passer le tableau 0,03, 0,03 et 0,05, ajouter 1 au tableau, puis prendre le produit cumulé du tableau. 

+ Le résultat est un tableau - 1,03, 1,06, 1,11 - qui est la croissance cumulative de l’investissement sur chaque période.
