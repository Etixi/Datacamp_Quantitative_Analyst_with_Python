####**Conversion de format avec datetimes**

+ *Avec datetimes, vous pouvez lire une chaîne avec un format et sortir une chaîne avec un format différent. Cela signifie que vous pouvez utiliser datetimes pour modifier le format des dates de chaîne. Les chaînes de format pour le mappage des dates et heures se trouvent dans strftime. Supposons qu'on vous demande de traiter la date du crash du mercredi noir britannique dans un nouveau format qui répond aux besoins de votre entreprise en matière de rapports.*

####**Instructions**
+ Créez une chaîne de format qui correspond à la date de la chaîne d'origine, donnée sous la forme org_text.
+ Créez une date/heure pour Black Wednesday et enregistrez-la sous black_wednesday.
+ Créez une chaîne de format qui correspond au nouveau format, "Mercredi 16 septembre 1992".
+ Créez une nouvelle date de chaîne en utilisant le nouveau format.

```python
org_text = "Sep 16 1992"

# Format string for original text
org_format = "%b %d %Y"

# Create datetime for Black Wednesday
black_wednesday = datetime.datetime.strptime(org_text, org_format)
print(black_wednesday)
```

```python
# New format: 'Wednesday, September 16, 1992'
new_format = "%A, %B %d, %Y"

# String in new format
new_text = black_wednesday.strftime(new_format)
print(new_text)
```

####**3. Que sont les opérations booléennes?**

+ **La logique booléenne consiste en des opérations effectuées entre des valeurs True ou False. Il existe trois de ces opérations : ET, également connue sous le nom de conjonction logique. OU, connu sous le nom de disjonction logique. Et NON, connu sous le nom de négation logique.**

####**4. Évaluation de l’objet**

+ **En Python, tous les objets sont évalués à True ou False pour les besoins de ces opérations. Les objets qui sont évalués comme False sont les constantes False et None. Zéro dans n’importe quel type numérique, y compris les entiers et les flottants. Tout ce qui a une longueur de zéro, y compris les chaînes vides, les listes et les dictionnaires. Les objets qui évaluent True incluent à peu près tout le reste.**

####**5. L’opérateur AND**
+ L’opérateur AND est représenté en Python par le mot-clé minuscule 'and'. 

+ Il compare deux éléments, et si les deux sont vrais, il donne la valeur true. Si l’un ou l’autre est faux, il est évalué à faux.

####**6. L’opérateur de la salle d’opération**

+ L’opérateur OR est représenté en Python par le mot-clé minuscule 'or'. Il donne la valeur true si l’un des éléments est true. 

+ Cela inclut le cas où les deux éléments sont vrais. Il n’est évalué à false que lorsque les deux éléments sont faux.

####**7. Court-circuit.**

+ Les opérations « et » et « ou » sont des opérations de court-circuit. 

+ Cela signifie que Python cessera d’évaluer les éléments qu’ils comparent dès que le résultat de l’opération sera déterminé. 

+ Par exemple, disons que nous avons deux méthodes, is_current() et is_investment(), qui renvoient toutes deux true ou false. 

+ Si nous utilisons ces méthodes comme entrées pour une opération AND et que la première méthode renvoie false, la seconde méthode ne sera pas appelée. 

+ De même si nous les utilisons avec l’opérateur OR, la seconde méthode ne sera évaluée que si la première retourne false.

####**8. L’opérateur NOT**


+ L’opérateur NOT est représenté par le mot-clé minuscule 'not'. 

+ Il renvoie le booléen opposé d’un élément qui le suit directement. 
+ Donc pas vrai retourne false et pas faux retourne vrai.

####**9. Ordre des opérations avec NOT**
+ L’opérateur NOT est inférieur dans l’ordre des opérations aux opérateurs non booléens. 
+ Par exemple, nous avons une opération d’égalité comparant deux booléens qui donne la valeur à false. 
+ Si nous plaçons l’opérateur NOT devant cela, cela annule le résultat de l’opération, pas le premier élément.

####**10. Évaluation de l’objet**

+ Étant donné que l’objet est évalué comme True ou False, vous pouvez les utiliser avec des opérateurs booléens. 

+ Par exemple, nous pouvons utiliser l’opérateur AND avec une chaîne.

####**11. Évaluation de l’objet**

+ l’opérateur OR avec une liste. Un vide dans cet exemple, qui donne la valeur False.

####**12. Évaluation de l’objet**

+ Ou l’opérateur NOT avec un dictionnaire. Le dictionnaire est vide, ce qui donne la valeur False, donc avec l’opérateur NOT, il donne la valeur True.

####**13. Retour d’objets**

+ **N’oubliez pas que les opérateurs AND et OR n’évaluent que le nombre d’éléments nécessaire. Ces deux opérateurs renvoient le dernier élément qu’ils évaluent. Ainsi, pour une opération AND avec deux éléments dont la valeur est True, telle que deux chaînes non vides, le deuxième élément est renvoyé. ###**Boucles For et while**

####**1. Boucles For et while**

####**2. Répétition d’un bloc de code**

+ ***Il y aura des situations où vous devrez répéter une opération sur plusieurs éléments. Par exemple, vous voudrez peut-être parcourir une liste de transactions et rechercher le symbole de sécurité pour chacune d’elles, ou peut-être vous arrêter à la première qui implique des titres Oracle.***

####**3. Boucles.**

+ **Python propose deux instructions composées pour de telles opérations, la boucle 'for' et la boucle 'while'. Ces instructions vous permettent d’exécuter un bloc de code plusieurs fois.**

####**4. Composants de l’instruction**

+ Tout comme l’énoncé « si », ces énoncés se composent de deux parties. Une instruction de contrôle et un bloc de code. 

+ L’instruction de contrôle détermine le nombre de fois que le bloc de code est exécuté.

####**5. Pour les boucles**

+ Les instructions de contrôle pour les boucles 'for' commencent par le mot-clé 'for', qui est suivi d’une variable, du mot-clé 'in' et d’une séquence. 
+ Comme dans les autres instructions de contrôle, il se termine par deux points. 

+ La séquence peut être n’importe quel objet que nous pouvons parcourir un élément à la fois, comme une liste, les touches d’un dictionnaire ou une chaîne.

####**6. Exemple de liste**


+ L’instruction 'for' parcourt en boucle les éléments de la séquence. Pour chaque itération, il affecte l’élément courant à la variable et exécute le bloc de code. 

+ Dans le cas d’une liste, la variable est affectée à chaque élément de la liste. 

+ La variable peut ensuite être utilisée dans le bloc de code, ici nous imprimons sa valeur.

####**7. Exemple de dictionnaire**

+ Dans le cas d’un dictionnaire, la variable est affectée à chaque clé. 

+ Ici, nous utilisons la clé dans le bloc pour imprimer les valeurs du dictionnaire.

####**8. Exemple de chaîne**
+ Pour une chaîne, la variable est affectée à chaque caractère individuellement. 

+ Dans cet exemple, nous imprimons chaque lettre séparément.

####**9. Tandis que les énoncés de contrôle**

+ L’instruction de contrôle d’une boucle « while » commence par le mot-clé « while », suivi d’une expression, puis de deux-points. 

+ Comme dans une instruction 'if', l’expression doit être évaluée à True ou False. 

+ La boucle 'while' exécutera le bloc de code à plusieurs reprises tant que l’expression est True.

####**10. Bien que l’exemple**
+ Ici, nous avons une boucle 'while' qui exécute un bloc de code encore et encore tant que la variable $x$ est inférieure à cinq. 

+ Le bloc imprime la valeur de $x$, puis l’augmente d’un.

####**11. Boucles infinies**
+ Il est important que votre boucle « while » ait un moyen de se terminer. 

+ Un bug courant pour les programmeurs débutants est d’utiliser une expression qui sera toujours vraie. 

+ Ici, nous n’incrémentons pas la variable, et donc l’expression de test est toujours vraie. 

+ Cela entraînera une boucle infinie de votre programme.

####**12. Sauter avec continuer**


+ L’instruction continue permet d’ignorer une itération de boucle. 

+ Il est généralement utilisé en combinaison avec une instruction « if » pour sauter les itérations qui répondent à certaines conditions. 

+ Dans cet exemple, nous ignorons l’itération lorsque la variable $x$ est définie sur le nombre 2.

####**13. Arrêt avec pause.**


+ L’instruction 'break' est utilisée pour sortir d’une boucle, sans terminer l’itération. 

+ Il est également généralement utilisé en conjonction avec une déclaration « si ». 

+ Dans cet exemple, nous effectuons une boucle à travers une série de transactions, mais arrêtons notre boucle la première fois que nous atteignons une transaction dont le symbole est « ORCL ». 

+ Notez que l’expression de contrôle dans ce cas est toujours True, donc si nous n’appelions pas 'break', ce serait une boucle infinie.

+ La médiane est calculée en ordonnant les éléments par magnitude et en prenant celui du milieu.

+ La variance nous donne des informations similaires à celles de l’écart-type, mais l’écart-type est plus facile à interpréter seul, tandis que la variance est plus utile pour une utilisation dans d’autres formules.
