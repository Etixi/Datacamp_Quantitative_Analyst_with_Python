####**3. Comment les pandas stockent les données**

+ **Pandas attribue un type de données différent à chaque colonne et stocke ces informations dans une propriété appelée 'dtype'**. 
+ **La dtype d’une colonne affecte la façon dont vous pouvez utiliser son contenu dans le calcul et la visualisation**. 
+ En particulier, les pandas distinguent quatre types principaux: 
  + **Le dtype 'object' est réservé aux colonnes contenant des données textuelles, ou un mélange de texte et de données numériques.** 
  
+ **Il existe deux types de données numériques** : 
  + **int64 est pour les colonnes contenant des nombres entiers, représentés en utilisant 64 bits de sorte que le plus grand nombre peut être 2 à la puissance de 64.** 
  + **float64 est le deuxième type de données numérique, réservé aux colonnes contenant des décimales ou des nombres entiers et des valeurs manquantes**. 

+ **Enfin, datetime64 dtype est destiné aux colonnes contenant des informations de date et d’heure.**

####**3. Tendance centrale**

+ ***Regardons d’abord les mesures de la tendance centrale, qui est une valeur considérée comme étant au centre, ou la plus typique, de vos données.*** 

+ ***Tout d’abord, vous pouvez calculer la moyenne: additionnez simplement vos données et divisez par le nombre d’observations. Deuxièmement, la médiane est la valeur qui partitionne vos données de sorte que 50 % des valeurs sont plus petites et 50 % plus grandes. Enfin, le mode est la valeur la plus fréquente. Il peut y avoir plusieurs modes si plusieurs valeurs se produisent avec la même fréquence. En pratique, cela se présente comme suit: Pour une distribution symétrique, ces trois mesures sont identiques : la moyenne est égale à la médiane, et cette valeur se produit également le plus fréquemment.**

####**4. Tendance centrale**

+ Cela change si les données sont faussées : 
  + les grandes valeurs s’étendant plus à droite donnent l’apparence d’une longue traîne. 
  + Dans ce cas, la moyenne est tirée vers les valeurs extrêmes de la longue traîne. 
  + Le mode représente toujours le seul pic de cette distribution, et la médiane se situe entre les deux.

####**5. Tendance centrale**

+ L’exemple suivant montre une distribution avec deux pics. 

+ Encore une fois, la moyenne se déplace plus loin vers les valeurs extrêmes, cette fois à gauche, et la médiane se situe entre la moyenne et le mode. 

+ Ce qu’il faut retenir, c’est que la moyenne est plus sensible aux valeurs aberrantes que la médiane, qui mesure mieux le point médian de vos données, à moins que les données ne suivent une distribution symétrique.

+ **Pour en savoir plus sur la dispersion de vos données, vous pouvez calculer la variance et l’écart-type. La variance est la somme de toutes les différences au carré entre vos données et la moyenne, divisée par le nombre d’observations moins 1. L’écart-type est la racine carrée de la variance. ####**4. Boxplot: quartiles et valeurs aberrantes**

+ Le diagramme en boîte est un graphique statistique classique basé sur les quantiles clés d’une distribution. 
+ La syntaxe est similaire aux tracés que vous avez vus dans la dernière vidéo. 

+ Le diagramme de boîte nécessite que vous fournissiez un DataFrame et que : 
  + l’étiquette d’une colonne catégorielle pour l’axe des abscisses, 
  + et Étiquette d’une colonne quantitative à afficher sur l’axe y. 
  
+ Après avoir fourni les données, le boxplot présente la répartition de la capitalisation boursière variable numérique pour chaque niveau de la variable catégorielle Secteur comme suit: 
  + La boîte commence au 1er et se termine au 3e quartile. 
  + La barre horizontale au centre de la boîte est le deuxième quartile, c’est-à-dire la médiane. 
  + L’intervalle interquartile est la différence entre le 3e et le 1er quartile, représentée par la longueur verticale de la boîte. 
  + Les deux moustaches s’étendant du haut et du bas ont une fois et demie la longueur de l’interquartile (mais limitées par 0 à l’extrémité inférieure). 
  + Les points de données en dehors de cette plage sont marqués comme valeurs aberrantes. 
  
+ Dans notre exemple, vous pouvez voir que les tailles d’entreprise varient considérablement à l’extrémité supérieure pour chaque secteur. 

+ Bien que la médiane soit similaire pour chaque secteur, le troisième quartile, la fourchette inter-quartile et le nombre de valeurs aberrantes diffèrent grandement.
