####**3. Qu’est-ce qu’un portefeuille**

+ **Investir dans le marché boursier est risqué car la valeur des cours des actions peut fluctuer fortement**. 
+ **En raison du risque inhérent à l’investissement dans une seule action, la plupart des gestionnaires de portefeuille professionnels diversifient leurs placements afin de réduire leur risque d’exposition à une seule société**. 
+ **En investissant dans une collection d’actions, également appelée portefeuille, les investisseurs peuvent généralement obtenir un meilleur compromis risque-rendement.**

####**4. Pourquoi avons-nous besoin d’une analyse de portefeuille**

+ Alors, pourquoi avons-nous besoin d’une analyse de portefeuille? 
+ Eh bien, lorsque vous investissez dans 1 action, vous pouvez facilement calculer le risque et le gain possible. 

+ Cependant, lorsque vous décidez de créer un portefeuille de 50 actions, par exemple, il devient difficile de comprendre même le gain ou le rendement possible.

####**5. Portefeuille versus fonds versus indice**

+ Mais d’abord, clarifions un peu la terminologie. 
+ ***Tout d’abord, un portefeuille est un ensemble d’actifs appartenant à un individu. Ces actifs peuvent être des actions, des obligations, des matières premières, des fonds, etc***. 
+ **Les gens utilisent parfois le fonds et le portefeuille de manière interchangeable.** 
+ **Cependant, un fonds est en fait un pool de placements géré par un gestionnaire de fonds professionnel**. 
+ **Les investisseurs individuels achètent des « parts » dans le fonds et le gestionnaire de fonds investit l’argent directement dans un portefeuille d’actions et d’obligations**. 
+ Enfin, **un indice est un ensemble théorique d’actifs, qui représente une partie du marché.** 
+ **Par exemple, l’indice composé Nasdaq représente les valeurs technologiques américaines.**

####**6. Investissement actif versus passif**

+ Parlons de différents styles d’investissement. 
+ Vous avez peut-être entendu parler de l’investissement passif. 
+ **Un investisseur passif vise à imiter les avoirs d’investissement d’un indice particulier et n’essaie donc pas de battre un indice de référence ou de maximiser les rendements**. 
+ **Un investisseur actif essaie généralement de battre un indice de référence ou de maximiser le rendement pour un niveau ou un risque donné.** 
+ **Les investisseurs actifs portent une attention particulière aux tendances du marché, aux changements dans l’économie, aux changements dans le paysage politique et aux facteurs susceptibles d’affecter des sociétés spécifiques, et utilisent ces informations pour acheter et vendre des actions.** 
+ Dans l’espace actif, les stratégies d’investissement long only tentent de surperformer un indice de référence, tandis que les hedgefunds à rendement absolu ne sont pas liés par un indice de référence et visent simplement un meilleur compromis risque-rendement possible.

####**7. Diversification**
+ **Lorsque vous investissez dans une seule action, des événements tels qu’un changement de direction, une faible performance financière, une performance sectorielle difficile, etc. feront tous fluctuer fortement votre investissement**. 
+ **En combinant les actions d’un portefeuille, vous pouvez réduire certaines de ces fluctuations, car les rendements positifs d’une action peuvent absorber les rendements négatifs d’autres.** 
+ **En combinant intelligemment des actifs avec différentes caractéristiques et différents profils de risque et de rendement, vous pouvez réduire le risque global de vos investissements. C’est ce qu’on appelle la diversification.**

####**8. Stratégies de portefeuille typiques**
+ Parlons des stratégies de portefeuille typiques. 
+ Le portefeuille à pondération égale, combine des actions et leur donne à toutes le même poids, comme son nom l’indique. 
+ Le portefeuille pondéré en fonction de la capitalisation boursière crée des pondérations individuelles basées sur la taille des sociétés, mesurée par la capitalisation boursière. 
+ **La capitalisation boursière est la valeur de l’entreprise calculée en multipliant les actions d’une société par le prix de marché actuel d’une action**. 
+ Enfin, un portefeuille optimisé risque-rendement a un compromis risque-rendement optimisé. Vous en apprendrez plus à ce sujet au chapitre 4.

###**Mesurer le risque d’un portefeuille**

####**1. Mesurer le risque d’un portefeuille**

####**2. Risque d’un portefeuille**

+ ***Lorsque vous investissez dans des actions, vous ne savez pas à l’avance quel sera votre rendement.*** 

+ ***Les prix montent et descendent, il y a donc un certain degré d’incertitude, ce qui implique que le rendement des actions est une variable aléatoire***. 

+ **La mesure dans laquelle les rendements réels sont répartis autour de leur valeur moyenne est appelée variance. Voici la formule officielle de variance. C’est une excellente indication du risque ou de la volatilité d’une action. Vous avez peut-être rencontré des variances dans votre classe de statistiques***.

####**3. Écart**

+ **Certaines actions ont une petite variance, ce qui signifie que leurs rendements sont toujours proches de la moyenne, comme la distribution des rendements ici en rouge**.

+ Ce dernier terme est en fait ce que nous appelons la covariance. 

+ Réécrivons donc la formule et insérons la covariance, au lieu de la corrélation multipliée par les variances.

+ Jetons un coup d’œil à ce portefeuille.

####**5. Coefficient d’asymétrie de Pearson**

+ **Le coefficient d’asymétrie de Pearson est calculé en regardant la différence entre la moyenne et la médiane, que vous avez vue sur la photo plus tôt**.

+ **Si le coefficient d’asymétrie est positif, cela signifie que la distribution a une bosse vers la gauche du milieu. Ce n’est pas bon pour investir, car une grande partie des rendements sera inférieure à la moyenne.**

  + Source : https://brownmath.com/stat/shape.htm

  + Source: Pimco

####**7. Interprétation du kurtosis**
Une kurtose élevée signifie qu’une grande partie de la variation provient d’événements extrêmes, ce qui correspond essentiellement aux queues grasses que nous avons vues plus tôt. Il est important de savoir qu’une distribution normale a un kurtosis de 3. Un kurtosis de moins de 3 signifie que les rendements sont plus concentrés autour de la moyenne, et moins dans les queues. Tout ce qui est au-dessus de 3 implique que vous avez des queues légèrement plus grosses, ce qui n’est pas souhaitable pour les investisseurs.

  + Source : https://brownmath.com/stat/shape.htm

####**2. Qu’est-ce qu’un facteur?**

+ **Les facteurs sont des facteurs de rendement larges et persistants.** 
+ **Tout comme les aliments sont constitués de nutriments comme les glucides et les protéines, les facteurs sont les éléments qui composent le rendement d’un portefeuille**. 

+ **Les actions de votre portefeuille présenteront certaines caractéristiques, telles que la volatilité, la taille, les solides fondamentaux financiers, etc**. 

+ **Ces caractéristiques détermineront le rendement de ces actions individuelles et, dans l’ensemble, formeront les facteurs de votre portefeuille**. 
+ **Les facteurs sont donc mesurés en examinant les caractéristiques combinées des actifs de votre portefeuille**. 

+ ***Supposons que vous ayez de nombreuses entreprises avec une grande taille mesurée par la capitalisation boursière dans votre portefeuille, cela signifie que vous avez une grande exposition au facteur de taille, ou autrement dit, que vous avez une forte corrélation avec le facteur de taille. Si ces grandes entreprises réussissent bien sur les marchés financiers actuels, votre portefeuille sera également performant.***

####**3. Facteurs dans les portefeuilles**

+ Nous pouvons généralement distinguer deux types de facteurs, les facteurs macroéconomiques et les facteurs de style. 

+ Les facteurs macroéconomiques sont les facteurs qui affectent les marchés financiers en général, la volatilité des taux d’intérêt, des taux de change, les changements dans les politiques gouvernementales telles que les tarifs commerciaux, etc. 

+ Les facteurs de style sont liés aux caractéristiques des actifs individuels, telles que la dynamique des prix, la volatilité des prix, le ratio comptable tel que le ratio valeur comptable / prix, etc.

####**4. Utilisation de modèles factoriels pour déterminer l’exposition au risque**

+ Tout comme vous voudriez changer les macronutriments dans votre alimentation lorsque vous avez un objectif spécifique à atteindre (par exemple, manger moins de glucides pour perdre du poids), vous pouvez également utiliser des facteurs comme blocs de construction pour créer votre portefeuille optimal, et ils sont utilisés de nombreuses façons. 

+ Dans le diagramme de gauche, les facteurs sont utilisés tactiquement.

+ En gérant stratégiquement et en combinant plusieurs facteurs dans un portefeuille, vous pouvez modifier le profil risque-rendement d’un portefeuille. Enfin, des facteurs sont également utilisés dans la gestion des risques, par exemple pour vérifier la diversification des portefeuilles.

####**3. Modèle Fama Français Multi Factor**


+ Voici à quoi ressemble le modèle Fama Français, il explique les rendements de votre portefeuille à l’aide de ces trois facteurs: 
  + MKT est le rendement du marché moins le taux sans risque. 
  + SMB signifie Small Minus Big: il mesure le rendement excédentaire des actions à faible capitalisation boursière par rapport à celles à capitalisation boursière plus importante. 
  - HML signifie High Minus Low. 
    + Il s’agit d’un facteur de valeur qui mesure le rendement excédentaire des actions de valeur, où les actions de valeur ont généralement un ratio valeur comptable / prix (B / P) plus élevé que les autres actions. Les bêtas nous indiquent comment ce facteur spécifique peut expliquer l’évolution des rendements de notre portefeuille.

####**4. Mise à jour du modèle de régression**


+ Vous avez peut-être remarqué que le modèle de Fama Français ressemble beaucoup à une régression linéaire. Voici un bref rappel; Une régression linéaire est un modèle qui tente d’ajuster une ligne droite à vos points de données. 

+ Il le fait en essayant de minimiser l’erreur globale, c’est-à-dire la distance entre les points bleus et la ligne rouge. 

+ Les bêtas du modèle Fama Français sont les mêmes que les coefficients bêta dans une régression linéaire : 
  + ils représentent la pente de la ligne et vous indiquent dans quelle mesure le rendement de votre portefeuille change si le facteur change de 1 unité.

####**5. Différence entre bêta et corrélation**

+ Vous vous demandez peut-être en quoi les bêtas sont différents des corrélations, alors soulignons quelques différences importantes ici. 

+ Lorsque le rendement factoriel change de 5 %, le bêta vous indique dans quelle mesure le rendement de votre portefeuille changera. 

+ La corrélation n’est pas si spécifique. Il vous indique en général dans quelle mesure la variation des rendements s’explique par les mouvements des facteurs. 

+ La corrélation est une mesure standardisée entre -1 et 1, et bêta ne l’est pas et sa taille a une signification. 

+ En outre, la corrélation n’a pas de direction, elle vous indique comment deux choses sont liées. Beta ne s’applique qu’à l’effet de $X$ sur $Y$, et non vice versa.

####**3. Qu’est-ce que l’optimisation de portefeuille?**


+ **Heureusement, il existe des modèles existants que vous pouvez utiliser pour vous aider à sélectionner des actions pour votre portefeuille. Harry Markowitz a mis au point Efficient Frontier Optimization en 1952, qui fournit un cadre solide pour combiner les actions d’un portefeuille. L’idée clé est qu’en combinant des actifs avec différents rendements et volatilités attendus, on peut décider d’une allocation mathématiquement optimale.**

####**4. Le problème d’optimisation : trouver des poids optimaux**

+ **L’optimisation efficace des frontières est un problème d’optimisation contraint, dans lequel vous essayez de minimiser la variance du portefeuille en définissant les pondérations**. 

+ La ligne du haut devrait vous sembler familière, c’est quelque chose que vous avez calculé plusieurs fois dans le chapitre 1, et c’est en fait la variance du portefeuille! Le reconnaissez-vous? 

+ Ses poids sont transposés, multipliés par la matrice de covariance sigma, multipliés par les poids à nouveau. 

+ Les conditions « sous réserve de » indiquent simplement que le rendement attendu, ou les pondérations multipliées par mu, doivent être au moins un certain niveau de rendement cible, mu star. 

+ Ensuite, il dit que les poids devraient aller jusqu’à 1, et qu’au moins certains des poids devraient être positifs.

####**5. La variation des rendements cibles conduit à la frontière efficiente**

+ **Lorsque vous modifiez l’étoile de retour cible mu dans le problème d’optimisation, les poids w qui forment la solution optimale, changent.** 

+ **Donc, en fait, la solution au problème d’optimisation de Markowitz, est une collection de portefeuilles, que nous appelons la frontière efficiente.** 

+ **La ligne du graphique représente cette frontière. Les points sont des portefeuilles individuels tracés avec leur profil de risque et de rendement respectif.** 

+ **La frontière efficiente est donc l’ensemble des portefeuilles optimaux qui offrent le risque le plus faible pour un niveau donné de rendement attendu et vice versa.**

```python
import pandas as pd
import requests
import numpy as np
from scipy.stats import percentileofscore as score
from statistics import mean
from math import floor

# CREATING S&P 500 SYMBOL LIST

sp500 = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
sp500_list = np.array(sp500[0]['Symbol'])
print(sp500_list[:20])

# EXTRACTING INTRADAY PRICES OF S&P 500 STOCKS

def get_intraday_prices(symbol):
    ticker = symbol
    iex_api_key = 'pk_32ef64b2003542b6a829b8f94831c789'
    url = f'https://cloud.iexapis.com/stable/stock/{ticker}/intraday-prices?token={iex_api_key}'
    df = requests.get(url).json()
    date = df[1]['date']
        
    time = []
    open = []
    high = []
    low = []
    close = []
    volume = []
    number_of_trades = []
    
    for i in range(len(df)):
        time.append(df[i]['label'])
        open.append(df[i]['open'])
        high.append(df[i]['high'])
        low.append(df[i]['low'])
        close.append(df[i]['close'])
        volume.append(df[i]['volume'])
        number_of_trades.append(df[i]['numberOfTrades'])
        
    time_df = pd.DataFrame(time).rename(columns = {0:'Time'})
    open_df = pd.DataFrame(open).rename(columns = {0:'Open'})
    high_df = pd.DataFrame(high).rename(columns = {0:'High'})
    low_df = pd.DataFrame(low).rename(columns = {0:'Low'})
    close_df = pd.DataFrame(close).rename(columns = {0:'Close'})
    volume_df = pd.DataFrame(volume).rename(columns = {0:'Volume'})
    number_of_trades_df = pd.DataFrame(number_of_trades).rename(columns = {0:'Number of Trades'})
     
    frames = [time_df, open_df, high_df, low_df, close_df, volume_df, number_of_trades_df]
    df = pd.concat(frames, axis = 1, join = 'inner')
    df = df.set_index('Time')
    return df
  
df = pd.DataFrame(columns = sp500_list)
for i in df.columns:
    try:
        df[i] = get_intraday_prices(i)['Close']
        print(f'{i} is successfully extracted')
    except:
        pass
    
df.to_csv('sp500.csv')

# IMPORTING THE EXTRACTED INTRADAY DATA

sp500 = pd.read_csv('sp500.csv').set_index('Time')
print(sp500.head())
```

```python
# CALCULATING DAY CHANGE OF STOCKS

dc = []
for i in sp500.columns:
    dc.append(sp500[i].pct_change().sum())
    
sp500_momentum = pd.DataFrame(columns = ['symbol', 'day_change'])
sp500_momentum['symbol'] = sp500.columns
sp500_momentum['day_change'] = dc

# CALCULATING MOMENTUM

sp500_momentum['momentum'] = 'N/A'
for i in range(len(sp500_momentum)):
    sp500_momentum.loc[i, 'momentum'] = score(sp500_momentum.day_change, sp500_momentum.loc[i, 'day_change'])/100
    
sp500_momentum['momentum'] = sp500_momentum['momentum'].astype(float)    
print(sp500_momentum.head())

top_picks = sp500_momentum.nlargest(10, 'momentum')['symbol'].reset_index().drop('index', axis = 1)
print(top_picks)

# BACKTEST

portfolio_val = 1000000
per_stock_val = portfolio_val/len(top_picks)

day_close = []
for i in top_picks['symbol']:
    data = sp500[i]
    day_close.append(data[-1])
    
backtest_df = pd.DataFrame(columns = ['selected_symbols', 'day_close', 'number_of_stocks', 'return', 'return_percentage'])
backtest_df['selected_symbols'] = top_picks['symbol']
backtest_df['day_close'] = day_close




for i in range(len(backtest_df)):
    backtest_df.loc[i, 'number_of_stocks'] = floor(per_stock_val/day_close[i])
    
returns = []
for i in top_picks['symbol']:
    ret = np.diff(sp500[i])
    ret = ret[~np.isnan(ret)]
    returns.append(round(sum(ret), 2))
    
backtest_returns = []
return_percentage = []
for i in range(len(backtest_df)):
    br = returns[i]*backtest_df.loc[i, 'number_of_stocks']
    rp = br/per_stock_val*100
    backtest_returns.append(round(br, 2))
    return_percentage.append(round(rp, 2))
backtest_df['return'] = backtest_returns
backtest_df['return_percentage'] = return_percentage

backtest_df
```
