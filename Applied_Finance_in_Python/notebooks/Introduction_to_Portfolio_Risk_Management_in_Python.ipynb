####**1. Moyenne, variance et distribution normale**

+ ***Étant donné que le risque financier peut simplement être considéré comme une mesure de l’incertitude des rendements futurs, il est de bonne pratique d’examiner la distribution naturelle des rendements afin de mieux comprendre le risque des investissements.***

####**2. Moments des distributions**

+ ***Les distributions de probabilité ont des propriétés communes, appelées moments, qui peuvent être analysées et comparées à d’autres distributions***. 

+ Le premier moment est **la moyenne, ou mu, qui est essentiellement le résultat moyen d’un échantillon aléatoire de la distribution**. 

+ Le deuxième moment est **la variance, qui est simplement l’écart-type, ou sigma, au carré. Il s’agit d’une mesure de la variabilité des résultats**. 

+ Les troisième et quatrième moments sont moins couramment utilisés. **Le troisième moment, l’asymétrie, est une mesure de « l’inclinaison » d’une distribution**,

+ tandis que le quatrième moment, **kurtosis, est une mesure importante de l’épaisseur de la queue d’une distribution**. Nous y reviendrons plus tard.

####**3. La distribution normale**

+ Il existe de nombreux types de distributions communes. Vous connaissez peut-être la loi normale gaussienne si vous avez déjà suivi un cours de statistiques.

####**4. La distribution normale standard**

La loi normale standard est un cas particulier de la loi normale lorsque sigma = 1 et mu = 0.

####**5. Comparaison avec une distribution normale**

+ Toutes les distributions normales ont tendance à avoir une asymétrie proche de 0 et une kurtose proche de 3. 

+ Les rendements financiers, en revanche, ont tendance à avoir une asymétrie positive et un kurtosis supérieur à 3. 

+ Cela signifie que les rendements financiers ont tendance à avoir une probabilité plus élevée de valeurs aberrantes et de rendements positifs qu’une distribution normale. Nous parlerons plus en détail de kurtosis et de skewness plus tard.

####**6. Comparaison avec une distribution normale**

####**Asymétrie et kurtosis**

####**1. Asymétrie et kurtosis**

+ La volatilité quotidienne et la moyenne vous donnent une bonne indication du risque et du rendement quotidiens, mais **l’asymétrie et la kurtose et la volatilité à l’échelle vous aideront à construire une vision plus holistique du risque**.

####**2. Asymétrie**

+ Le troisième moment, l’asymétrie peut être considéré comme une mesure de la mesure dans laquelle une distribution penche vers la gauche ou la droite. 

+ L’inclinaison négative est une courbe inclinée vers la droite, tandis que l’inclinaison positive est une courbe inclinée vers la gauche. 

+ En finance, vous auriez tendance à vouloir une asymétrie positive, avec une probabilité plus élevée de rendements significativement bons du côté droit de la distribution, et une distribution comprimée et prévisible à gauche des rendements négatifs.

####**3. Asymétrie en Python**

+ Pour calculer l’asymétrie d’une distribution de retour, appelez simplement la fonction skew() après avoir importé l’inclinaison à partir des statistiques de points de scipy. 

+ Assurez-vous de supprimer toutes les valeurs NA à l’aide de la méthode dropna(). 

+ Une asymétrie supérieure à 0 indique une non-normalité possible, ce que vous pouvez encore une fois vous attendre à trouver dans les rendements financiers.

####**4. Kurtosis**

+ La kurtose est une mesure de l’épaisseur des queues d’une distribution, qui peut être utilisée comme indicateur de la probabilité de valeurs aberrantes. 

+ Si vous vous souvenez, les distributions normales ont tendance à avoir un kurtosis près de 3. 

+ La plupart des rendements financiers sont leptokurtiques, ce qui signifie simplement qu’ils ont tendance à avoir un excès positif de kurtose, ou kurtosis supérieur à 3. 

+ Étant donné que la kurtose est si souvent comparée à une distribution normale, de nombreuses fonctions en Python retourneront automatiquement l’excès de kurtose, qui est essentiellement l’échantillon kurtosis moins 3, ce qui aide à démontrer si la probabilité de valeurs aberrantes est supérieure ou inférieure à une distribution normale. 

+ Si l’excès de kurtose est supérieur à 0, le kurtosis est supérieur à une distribution normale.

####**5. Excès de kurtosis en Python**

+ Dans SciPy, par exemple, la fonction kurtosis calcule en fait l’excès de kurtose. Si vous vouliez calculer la kurtose de l’échantillon réel, vous auriez en fait besoin d’ajouter 3 au résultat. 

+ Mais dans la plupart des cas, vous allez être intéressé par l’excès de kurtosis de toute façon, donc cette fonctionnalité est bien tant que vous en êtes conscient. 

+ En finance, un excès élevé de kurtosis est une indication de risque élevé. Lorsque d’importants mouvements de rendement se produisent souvent, cela peut être une très mauvaise chose pour votre portefeuille s’il évolue dans la mauvaise direction. 

+ On dit que les distributions élevées de kurtosis ont des « queues épaisses », ce qui signifie que les valeurs aberrantes, telles que les rendements négatifs et positifs extrêmes, sont plus courantes.

####**6. Test de normalité en Python**

+ Avant de poursuivre, permettez-moi d’ajouter brièvement un dernier outil à votre arsenal. **Si le kurtosis d’une distribution est supérieur à 3 et que l’asymétrie est non nulle, les données sont très probablement anormales**. 

+ ***Mais que se passe-t-il si les valeurs sont proches, mais pas tout à fait normales? Vous pouvez utiliser le test statistique de Shapiro-Wilk pour estimer la probabilité que les données soient normalement distribuées.*** 

+ ***L’hypothèse nulle du test de Shapiro-Wilk est que les données sont normalement distribuées, et lorsque la valeur de p renvoyée est inférieure ou égale à 0-point-05, cela signifie que vous pouvez rejeter en toute sécurité l’hypothèse nulle et supposer que les données sont non normales***.

###**Corrélation et covariance**

####**1. Corrélation et covariance**

+ La théorie moderne du portefeuille enseigne que vous pouvez réellement construire un portefeuille d’actifs qui présente moins de risques que n’importe lequel des actifs sous-jacents seuls. 

+ Comment est-ce possible? En bref, tout est question de corrélation et de covariance.

####**2. Corrélation de Pearson**

+ **La corrélation de Pearson est une mesure de la corrélation linéaire entre deux variables, dans notre cas, entre les rendements de deux actions**. 

+ La corrélation de Pearson varie de -1 à 1 où 1 est une corrélation positive parfaite, -1 serait une corrélation négative parfaite et 0 serait aucune corrélation.

####**3. Corrélation de Pearson**

+ Vous pouvez calculer le coefficient de corrélation pour chaque paire d’actions de votre portefeuille, ce qui donne une matrice de corrélations. 

+ Je ne vais pas vous ennuyer avec la formule, puisque vous allez simplement utiliser Python pour calculer facilement la matrice de corrélation de Pearson.

####**4. Matrice de corrélation en Python**

+ En supposant que vous ayez un DataFrame pandas de rendements boursiers, tout ce que vous avez à faire est d’appeler **la méthode dot corr()** sur le DataFrame. 

+ Vous remarquerez que les diagonales de la matrice sont toujours égales à 1. 

+ En effet, toute variable ou actif sera toujours parfaitement corrélé avec lui-même, bien sûr.

####**5. Écart-type du portefeuille**

+ Nous pouvons maintenant utiliser la corrélation dans une formule pour calculer l’écart-type du portefeuille pour un portefeuille à deux actifs. 

+ Mais si vous avez plus d’actifs dans votre portefeuille, cette formule deviendra très compliquée très rapidement. 

+ C’est là que la matrice de covariance sera utile.

####**6. La matrice de covariance**

+ La corrélation est en fait une mesure normalisée de la covariance. 

+ La covariance mesure la variabilité conjointe de deux variables aléatoires et, en finance, la matrice de covariance est souvent utilisée à des fins d’optimisation du portefeuille et de gestion des risques. 

+ Il existe de nombreuses façons de dériver une matrice de covariance, mais bien sûr, pour ce cours, nous utiliserons simplement une fonction Python pratique!

####**7. La matrice de covariance en Python**

+ Tout ce que vous avez à faire est d’appeler **la méthode dot cov()** sur le DataFrame des rendements boursiers quotidiens.

####**8. Annualisation de la matrice de covariance**

+ Maintenant, vous vous souvenez peut-être que la variance d’une action évolue linéairement avec le temps, ou la volatilité avec la racine carrée du temps. 

+ Dans ce cas, puisqu’il s’agit d’une matrice variance-covariance, tout ce que vous avez à faire pour annualiser la matrice de covariance est de la multiplier par le nombre de jours de bourse dans une année, soit 252.

####**9. Écart-type du portefeuille à l’aide de la covariance**

####**3. Rapport de Sharpe**
+ ***Afin de comparer le rendement du portefeuille, vous voudrez tenir compte des rendements ajustés au risque***. 

+ ***Le ratio de Sharpe est une mesure utile qui est citée assez fréquemment dans la finance à cette fin***. 

+ ***Le ratio de Sharpe mesure le rendement auquel un investisseur peut s’attendre pour chaque unité de risque supplémentaire et peut donc être utilisé pour comparer des portefeuilles présentant différents niveaux de risque***. 

+ Cette formule suppose un taux de rendement sans risque, mais pour ce chapitre, nous pouvons supposer que cette valeur est égale à 0.

+ Dans cet exercice, vous utiliserez la formule simple suivante impliquant la covariance et la variance par rapport à un portefeuille de marché de référence :

$$
 \beta_P = \frac{Cov(R_P, R_B)}{Var(R_B)} 
$$

+ $\beta_P$: Bêta du portefeuille
+ $Cov(R_P, R_B)$: La covariance entre le portefeuille (P) et l'indice boursier de référence (B)
+ $Var(R_B)$: La variance de l'indice boursier de référence

+ ***Il existe de nombreuses façons de modéliser les rendements boursiers, mais le modèle d'évaluation des actifs financiers, ou CAPM, est l'un des plus connus*** :

 $$E(R_{P}) - RF = \beta_{{P}}(E(R_{M})-RF)$$ 

+ $E(R_{P}) - RF$: L'excès de rendement attendu d'une action ou d'un portefeuille P
+ $E(R_{M}) - RF$ : L'excès de rendement attendu du portefeuille de marché large B
+ $RF$: Le taux sans risque régional
+ $\beta_{{P}}$: Bêta du portefeuille, ou exposition, au portefeuille de marché large B

####**2. Le modèle Fama-Français à 3 facteurs**

+ Le modèle Fama-Français 3 facteurs est peut-être l’un des modèles les plus largement utilisés pour la gestion de portefeuille en finance, et a été une véritable percée pour la théorie moderne du portefeuille à l’époque. 

+ ***Développé par Eugene Fama et Kenneth Français au début des années 1990, ce modèle étend le modèle CAPM en ajoutant le SMB, ou facteur de taille, ainsi que le HML, ou facteur de valeur***. 

+ ***L’acronyme de valeur, HML, signifie élevé moins faible, tandis que l’acronyme de taille, SMB, signifie petit moins grand***. 

+ ***C’est parce que les petites actions ont tendance à surperformer les grandes actions, de sorte que les petits rendements d’actions moins les rendements de grandes actions sont essentiellement la prime de petite taille***. 

+ La valeur par rapport à la croissance, ou élevée moins faible, en revanche, est plus cyclique. 

+ Par exemple, pendant la crise, les actions de valeur ont surperformé, mais pendant les périodes haussières, les actions de croissance surperforment parfois.

Les investisseurs devraient théoriquement être récompensés pour avoir été exposés à des actions à petite capitalisation plutôt qu’à des actions à grande capitalisation, ainsi que pour avoir été exposés à des actions en difficulté ou de valeur au lieu des actions de croissance populaires et premium. Plus de risques, plus de récompenses.

####**7. L’alpha et l’hypothèse du marché efficient**
+ Tout ce qui ne peut pas être expliqué par les facteurs bêta, de taille ou de valeur dans le modèle est appelé alpha, qui est essentiellement un terme d’erreur. 

+ L’alpha positif est maintenant communément interprété comme une surperformance due à la compétence, à la chance ou au timing. 

+ Bien sûr, pour chaque fonds avec un alpha positif, il existe un autre fonds avec un alpha négatif puisque la somme pondérée de tout l’alpha sur un marché doit être de 0. 

+ En effet, la somme pondérée des rendements de tous les investisseurs est simplement égale au portefeuille du marché. 

+ Certains économistes croient en l’hypothèse du marché efficient, qui stipule essentiellement que le prix du marché est de toutes les informations. 

+ Par conséquent, tout « alpha » perçu est simplement le résultat d’un facteur manquant dans un modèle de tarification économique plus complexe.

###**Estimation du risque extrême**

####**1. Estimation du risque extrême**
+ ***Un élément clé de la gestion du risque de portefeuille est d’être capable de définir le risque au-delà de simples mesures telles que l’écart-type, la variance et même la kurtose***. 

+ ***Plus important est la capacité de quantifier les pertes potentielles au niveau d’une seule action et d’un seul portefeuille***.

####**2. Estimation du risque extrême**


+ ***Le risque extrême est essentiellement le risque de résultats extrêmes dans la queue de la distribution du rendement, en particulier dans la queue gauche, ce qui signifie un rendement extrêmement négatif***. 

+ Dans ce chapitre, vous couvrirez plusieurs méthodes d’estimation du risque extrême, y compris les pertes historiques, deux méthodes différentes pour **la valeur à risque et la valeur conditionnelle à risque, et enfin - la simulation de Monte Carlo**.

####**3. Prélèvement historique**
+ ***Le drawdown historique est un moyen d’estimer le pourcentage de perte à partir du point le plus élevé***. 

+ En d’autres termes, jusqu’où êtes-vous tombé de votre meilleur moment de l’histoire? 

+ Idéalement, les bons investissements auraient peu ou pas de retrait, car ils augmenteraient simplement de manière constante au fil du temps, mais en réalité, vous pouvez vous attendre à ce que des retraits se produisent assez fréquemment dans la plupart des portefeuilles.

####**4. Drawdown historique en Python**


+ Compte tenu de la formule de la diapositive précédente, il est relativement simple de calculer le drawdown historique en Python en utilisant les fonctions maximum et cumulative pour calculer le maximum cumulé dans le temps, ce qui est essentiel à la formule. 

+ Tout d’abord, appelez np point maximum point accumuler sur vos rendements cumulés pour obtenir votre maximum de course au fil du temps. 

+ Mais vous n’êtes intéressé que par les pertes, alors ensuite, définissez le maximum de course à être dépassé par 1, qui est votre point de départ. 

+ Enfin, divisez les rendements cumulés par le maximum cumulé et soustrayez 1 pour centrer la série sur 0. C’est votre drawdown, et comme vous pouvez le voir, il ne dépassera jamais 0.

####**5. Valeur historique à risque**
+ La valeur à risque, souvent appelée VaR, est un moyen d’estimer le risque d’un mouvement de prix négatif en une seule journée. 

+ La VaR a toujours un quantile, ou un pourcentage qui lui est attaché. 

+ Par exemple, une VaR 95 de -2 points 3% signifie que dans les 5% les pires scénarios, mes pertes dépasseront -2 points 3%. 

+ Vous pouvez y penser dans l’autre sens - je peux être certain à 95% que mes pertes ne dépasseront pas -2 points 3% dans une journée donnée.

####**6. Valeur historique à risque en Python**

+ Vous pouvez utiliser la fonction de centile np point pour calculer facilement le quantile à 95 % d’une distribution. 

+ Mais dans cet exemple, vous voulez passer 100 moins 95, ce qui n’est que 5. 

+ Si vous vouliez atteindre le 90e centile, vous passeriez 100 moins 90, ce qui correspond à 10. 

+ Notez simplement que cette fonction attend des entiers au lieu de décimales, par exemple 10 au lieu de 0,10.

####**7. Manque à gagner historique prévu**
+ ***La valeur conditionnelle à risque, également connue sous le nom de CVaR, est une mesure de la valeur attendue des pertes dans les pires scénarios de 1 moins $x$ pour cent, d’où le nom alternatif « déficit prévu » est dérivé. Dans cet exemple, CVaR(95) = -2.5 % signifie que dans les 100 à 95 pires cas, ou les 5 % les pires cas, les pertes seront en moyenne de -2.5 %.*** 

+ C’est essentiellement la même chose que de prendre la moyenne des pertes dépassant le niveau de VaR(95). Tout comme la VaR, les CVaR 95, 99 et 99 points 9 sont courants.

####**8. Déficit historique attendu en Python**

+ Par exemple, en Python, vous pouvez simplement calculer $VaR(95)$ comme auparavant, puis sous-définir les retours, en prenant la moyenne de tous les retours inférieure ou égale à la $aR(95)$. 

+ ***Cela signifie que votre CVaR sera toujours pire que votre VaR du même quantile***.

+ *Utilisez la fonction np.random.normal() pour modéliser les mouvements de marche aléatoires de l'ETF pétrolier USO avec un rendement moyen quotidien constant (mu) et une volatilité quotidienne moyenne (vol) au cours de T jours de négociation.*

####**Instructions**

+ Définissez le nombre de jours simulés (T) égal à 252 et le cours initial de l'action (S0) égal à 10.
+ Calculez T valeurs normales aléatoires à l'aide de np.random.normal(), en passant mu et vol, et T comme paramètres, puis en ajoutant 1 aux valeurs et affectez-le à rand_rets.
+ Calculez la marche aléatoire en multipliant rand_rets.cumprod() par le cours initial de l'action et affectez-la à forecasted_values.

```python
# Set the simulation parameters
mu = np.mean(StockReturns)
vol = np.std(StockReturns)
T = 252
S0 = 10

# Add one to the random returns
rand_rets = np.random.normal(mu, vol, T) + 1

# Forecasted random walk
forecasted_values = S0*rand_rets.cumprod()

# Plot the random walk
plt.plot(range(0, T), forecasted_values)
plt.show()
```

####**Simulations de Monte-Carlo**


+ ***Les simulations de Monte-Carlo sont utilisées pour modéliser un large éventail de possibilités.***

+ ***Monte-Carlos peut être construit de nombreuses manières différentes, mais toutes impliquent de générer un grand nombre de variantes aléatoires d'un modèle donné, permettant d'analyser une large distribution de chemins possibles. + Appelez la fonction de traçage pour chaque itération à l'aide de la fonction plt.plot(), en passant la plage de valeurs T (range(T)) comme premier argument et les valeurs_prévues comme deuxième argument.

```python
# Loop through 100 simulations
for i in range(100):

    # Generate the random returns
    rand_rets = np.random.normal(mu,vol,T) + 1
    
    # Create the Monte carlo path
    forecasted_values = S0*(rand_rets).cumprod()
    
    # Plot the Monte Carlo path
    plt.plot(range(T), forecasted_values)

# Show the simulations
plt.show()
```

+ ***Les valeurs de rendement et les chemins de Monte-Carlo peuvent être utilisés pour l'analyse de tout, allant des modèles d'évaluation des options et de la couverture à l'optimisation du portefeuille et aux stratégies de négociation.***
