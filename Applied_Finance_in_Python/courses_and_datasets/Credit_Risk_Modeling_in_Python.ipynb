###**Comprendre le risque de crédit**

####**1. Comprendre le risque de crédit**


+ Bonjour, je m’appelle Michael Crabtree et je suis data scientist chez Ford Motor Company. 

+ Je vais vous montrer quelques concepts et techniques pour la modélisation du risque de crédit en utilisant Python.

####**2. Qu’est-ce que le risque de crédit?**

+ ***Qu’est-ce que le risque de crédit exactement? Le risque de crédit est le risque que quelqu’un qui a emprunté de l’argent ne rembourse pas tout***. 

+ ***Considérez ce risque comme la différence entre prêter de l’argent à une personne et acheter une obligation d’État***. 

+ Avec les obligations d’État, il est presque garanti d’être remboursé, mais pas lorsque vous prêtez de l’argent à des gens. 

+ ***Un prêt est en défaut lorsque l’organisme prêteur est raisonnablement certain que le prêt ne sera pas remboursé***. 

+ ***Nous utiliserons des modèles d’apprentissage automatique pour le déterminer***.

####**3. Qu’est-ce que le risque de crédit?**

+ Prenons cet exemple : 
  + nous avons prêté 300 dollars à quelqu’un qui a effectué deux paiements, mais pas le paiement final. 
+ C’est à ce stade que nous considérons que le prêt est en défaut. 

+ Prédire cela à l’avance est utile pour nous permettre d’estimer la perte attendue.

####**4. Perte attendue**

+ ***La perte attendue est le montant que l’entreprise perd à la suite du défaut de paiement d’un prêt***. 

+ ***La perte attendue est un calcul simple des trois composantes suivantes***. 

+ ***La probabilité de défaut, qui est la probabilité que quelqu’un fasse défaut sur un prêt. L’exposition au défaut qui est le montant impayé au moment du défaut. Et la perte donnée par défaut qui est le rapport de l’exposition par rapport à toute récupération de la perte.***

+ D’après notre exemple, les 100 dollars qui nous étaient dus sont notre exposition, et si nous vendons cette dette pour 20 dollars, notre perte en cas de défaut serait de 80%. 

+ La formule de la perte attendue est la probabilité de défaut multipliée par l’exposition au défaut et la perte en cas de défaut. 

+ Ce cours se concentrera sur la probabilité de défaut.

####**5. Types de données utilisées**
+ Pour modéliser la probabilité de défaut, nous avons généralement deux principaux types de données disponibles. 

+ Le premier est les données de demande, qui sont des données directement liées à la demande de prêt, comme la qualité du prêt. 

+ La seconde est les données comportementales, qui décrivent le bénéficiaire du prêt, comme la durée de l’emploi.

+ **Normalement, vous pouvez penser que des données manquantes sont manquantes, mais ce n’est pas la seule façon dont les données peuvent être manquantes**. 

+ **Les données peuvent être manquantes lorsqu’il existe des valeurs nulles à la place des valeurs réelles**. 

+ Il peut également s’agir d’une chaîne vide au lieu d’une vraie chaîne. 

+ Pour ce cours, nous ferons référence aux données manquantes lorsque des valeurs spécifiques ne sont pas présentes, et non lorsque des lignes entières de données sont manquantes. 

+ N’importe laquelle des colonnes de nos données peut contenir des valeurs manquantes. 

+ Si nous voyons une ligne de données avec des valeurs manquantes dans une trame de données Pandas, cela ressemblera à ceci. Avis pour la durée de l’emploi, nous voyons NAN, ou pas un nombre, au lieu d’une valeur.

####**3. Similitudes avec les valeurs aberrantes**
+ ***Un problème avec les données manquantes est similaire aux problèmes causés par les valeurs aberrantes en ce sens qu’il a un impact négatif sur les performances du modèle prédictif***. 

+ Cela peut biaiser notre modèle de manière imprévue, ce qui peut affecter la façon dont nous prévoyons les défauts. 

+ Cela pourrait nous amener à prédire un grand nombre de défauts qui ne sont pas réellement des défauts parce que le modèle est biaisé en faveur des défauts. 

+ En outre, de nombreux modèles d’apprentissage automatique en Python n’ignorent pas automatiquement les valeurs manquantes et lancent souvent des erreurs et cessent la formation.

#####**2. Arbres de décision**

+ **Alors, qu’est-ce qu’un arbre de décision?**

+ **Ce sont des modèles d’apprentissage automatique qui utilisent les décisions comme étapes d’un processus pour éventuellement identifier le statut de notre prêt**. 

+ Bien qu’ils produisent des prédictions similaires aux régressions logistiques, ils ne sont pas structurés de la même manière. 

+ Voici un exemple d’arbre de décision simple. 

+ La première boîte, ou nœud, a décidé de diviser les données en deux groupes. 

+ Ceux dont l’ancienneté est supérieure à 10 ans et ceux dont l’ancienneté est inférieure à 10 ans. 

+ Ensuite, il utilise l’intention de prêt médical de la même manière. 

+ Les résultats de ces scissions sont des décisions oui et non qui mènent éventuellement à un statut de prêt prédit de défaut ou de non-défaut.

####**3. Arbres de décision pour l’état du prêt**

+ Jetons un coup d’œil à un exemple simple d’arbre de décision sur les données de prêt lorsque ce que nous prédisons est toujours des défauts de paiement.

+ La zone ombrée en rouge est ce que notre modèle a prédit par défaut. 

+ Bien qu’il ait prédit correctement toutes les valeurs par défaut, il a prédit deux valeurs par défaut non définies par défaut.

#####**4. Impact de l’arbre décisionnel**

+ Quelles en sont les conséquences? 

+ Disons que ces deux prêts valaient 1500 et 1200 au moment où nous avons prédit leur statut.

+ En raison du modèle, notre perte est de 2200 dollars pour seulement deux prêts!

####**5. Une forêt d’arbres**
+ XGBoost n’utilise pas un seul arbre de décision, mais un grand nombre d’entre eux dans ce que l’on appelle un ensemble grâce à une méthode appelée amplification de gradient. 

+ Chaque arbre de l’ensemble est comme celui que nous venons de voir, et est un prédicteur faible. 

+ Jetez un coup d’œil à cet exemple. 

+ Les deux premières cases à gauche représentent deux modèles individuels différents. 

+ Chacun d’eux prédit les valeurs par défaut, mais ils prédisent également certains non-défauts comme défauts. 

+ Cependant, lorsque nous utilisons l’amplification de gradient avec XGBoost, nous obtenons la boîte à droite qui combine les deux modèles faibles. Dans cet exemple, le modèle boosté prédit correctement tous les prêts.

+ ***C’est là que nous utilisons la validation croisée pour tester nos modèles afin de voir comment ils pourraient fonctionner après leur implémentation.***

####**2. Principes de base de la validation croisée**

+ Mais qu’est-ce que la validation croisée ? 

+ En termes simples, il s’agit d’une méthode permettant d’entraîner et de tester un modèle plusieurs fois sur les mêmes données. 

+ Nous ne pouvons pas créer plus de données de prêt pour nous aider à développer notre modèle, mais nous pouvons utiliser la validation croisée pour simuler la performance de notre modèle sur les nouvelles données de prêt avant qu’elles n’arrivent. 

+ ***Dans XGBoost, pour utiliser la validation croisée, vous devez créer un objet spécialisé appelé dmatrix, qui est juste une façon différente de stocker les données d’entraînement.***

+ Nous devons également utiliser l’arrêt précoce pour maintenir la robustesse des modèles. 

+ Cela indique à la validation croisée de s’arrêter lorsque le score du modèle ne s’est pas amélioré après un nombre défini d’itérations.

####**3. Fonctionnement de la validation croisée**

+ Voici comment fonctionne la validation croisée. 

+ Ce que nous faisons, c’est prendre l’ensemble des données d’entraînement et en créer des pièces, appelées plis. 

+ Tous les plis sauf un sont utilisés pour l’entraînement, et le pli restant est utilisé comme une sorte d’ensemble de test miniature. 

+ Une fois les tests terminés sur tous les plis, le modèle est testé par rapport à l’ensemble de test réel. Nous avons créé 5 plis. 

+ Maintenant, ce qui se passe, c’est que le modèle est entraîné sur 4 des plis et testé par rapport au pli final. 

+ Ce processus se répète à travers 5 fractionnements afin que chaque pli soit utilisé pour les tests au moins une fois. 

+ Une fois cela fait, les paramètres sont moyennés sur chaque session d’entraînement, puis le modèle est finalement testé par rapport à l’ensemble de test d’origine.

  + https://scikit-learn.org/stable/modules/cross_validation.html


####**4. Configuration de la validation croisée dans XGBoost**
+ Voici comment nous utilisons la validation croisée à partir de xgboost. 

+ Tout d’abord, nous définissons le nombre de plis. 

+ Ensuite, nous définissons le nombre d’itérations que nous autoriserons avant l’arrêt des simulations. 

+ Ensuite, nous créons un dictionnaire de paramètres. Le paramètre logistique binaire que nous avons créé indique à xgboost que nous voulons prédire un 0 ou un 1 pour le statut du prêt. 

+ La mesure de performance ici est la zone sous la courbe. Il s’agit de la même mesure que nous avons utilisée sur les modèles de régression logistique.

+ Enfin, nous appelons la fonction cv et passons les données longues avec tous les paramètres du dictionnaire.

###**Déséquilibre de classe dans les données sur les prêts**

####**1. Déséquilibre de classe dans les données sur les prêts**

+ Nous venons d’utiliser la validation croisée pour vérifier la robustesse de notre modèle. Parlons davantage de l’impact des données sur la robustesse du modèle.

####**2. Pas assez de valeurs par défaut dans les données**

+ Dans nos données de crédit, les valeurs pour le statut du prêt sont nos classes. 

+ En regardant le nombre de valeurs, nous voyons que le nombre de valeurs par défaut et de non-valeurs par défaut n’est pas égal. 

+ Dans les données d’entraînement, les valeurs par défaut représentent 22%. 

+ Il y a donc beaucoup plus de non-défauts que de défauts. ***C’est un déséquilibre de classe, qui est un problème.***

####**3. Fonction de perte de modèle**
+ ***Nos modèles arborescents utilisent une fonction de perte appelée perte logarithmique***. 

+ ***Notre modèle voudra prédire les deux classes avec précision, car cela conduit à un meilleur score global***. 

+ Ici, nous avons une valeur par défaut et une autre par défaut. 

+ Chacun a une probabilité prédite très éloignée du statut réel du prêt. 

+ Avec la fonction log-loss, les deux aboutissent à la même valeur. 

+ Donc, pour notre modèle, chacun est également mauvais. Le problème est que, pour les prêts, un défaut prévu comme étant un non-défaut est beaucoup plus coûteux.

####**4. Le coût du déséquilibre**

+ Disons que nous avons deux prêts pour mille dollars chacun. 

+ L’un est une valeur par défaut prédite par défaut et l’autre est une valeur par défaut prédite par défaut. Ici, chacun a un profit potentiel de 10 dollars. 

+ Si nous refusons un prêt parce que nous pensons que c’est un défaut alors que ce n’est pas le cas, nous manquons 10 dollars. 

+ Si nous acceptons un prêt parce que nous pensons que c’est un non-défaut alors qu’il s’agit d’un défaut, nous pourrions perdre mille dollars. 

+ Pour le modèle, chacun d’eux est également mauvais. Pour notre portefeuille, ce n’est certainement pas le cas.

####**5. Causes du déséquilibre**
+ ***Ce qui cause le déséquilibre de classe peut être plusieurs choses différentes***. 

+ ***La première est qu’il pourrait y avoir des problèmes avec les données elles-mêmes.*** 

+ ***Peut-être que les données d’entraînement n’ont pas été échantillonnées correctement***. 

+ Pour nous, le déséquilibre est tout à fait normal. 

+ Les gens sont incités à ne pas faire défaut sur leurs prêts. 

+ Par exemple, moins ils font défaut souvent, plus leur cote de crédit augmente et ils sont en mesure d’acheter plus de choses.

####**6. Faire face au déséquilibre de classe**

+ ***Alors, comment pouvons-nous faire face au déséquilibre de classe? Une façon consiste à recueillir plus de données***. 

+ Cependant, s’il reste normal que les gens ne fassent pas défaut sur leurs prêts, il est peu probable que le pourcentage de défauts de paiement change. 

+ Une autre façon est de pénaliser la classification différemment. 

+ Cela aurait fait peser davantage les modèles sur les défauts prédits de manière inexacte. 

+ Cependant, cela peut nécessiter un réglage et une maintenance fréquents des paramètres du modèle pendant toute la durée de vie du modèle. 

+ Nous utiliserons la méthode la plus simple, qui consiste à échantillonner les données d’entraînement différemment.

####**7. Stratégie de sous-échantillonnage***
+ ***Le type d’échantillonnage que nous allons effectuer est appelé sous-échantillonnage.*** 

+ ***Ce que nous allons faire, c’est prendre un échantillon aléatoire de non-défauts et le combiner avec nos défauts***. 

+ Imaginez que nous ayons 100 prêts dont 80% ne sont pas en défaut de paiement. Nous échantillonnerons au hasard seulement 20 de nos valeurs non par défaut et les combinerons avec notre ensemble de 20 valeurs par défaut. 

+ Avec cela, nous avons un ensemble de formation équilibré de 20 valeurs par défaut et 20 non-défauts.

####**8. Combinaison des ensembles de données fractionnés**
+ Voici donc la façon dont nous sous-échantillonnons nos données d’entraînement. 

+ Nous avons d’abord concaténé nos ensembles d’entraînement $X$ et $y$ parce que nous les avons séparés pour notre modèle, mais nous devons maintenant modifier les données d’entraînement dans leur ensemble.

+  Ensuite, nous stockons les comptes pour les valeurs par défaut et les non-défauts. 

+ Après cela, nous créons deux nouveaux ensembles de données. Un avec seulement des valeurs par défaut et un avec seulement des non-défauts.

####**9. Sous-échantillonnage des éléments non par défaut**
+ Cela fait, nous échantillonnons au hasard nos non-défauts de paiement pour qu’ils aient le même nombre de prêts que nos défauts de paiement. 

+ Ensuite, nous concatérons les deux ensembles de données ensemble, et nous avons un ensemble d’entraînement équilibré!

####**3. Seuils et taux d’acceptation**
+ Nos modèles ont déjà prédit les probabilités de défaut, et nous pouvons utiliser ces probabilités pour calculer le seuil. 

+ Étant donné que le seuil est utilisé pour déterminer ce qui est un défaut ou non, il peut également être utilisé pour approuver ou refuser de nouveaux prêts au fur et à mesure qu’ils arrivent. 

+ Par exemple, supposons que notre ensemble de tests soit un nouveau lot de nouveaux prêts. 

+ Avant de calculer le seuil, nous devons comprendre un concept connu sous le nom de taux d’acceptation.

####**4. Comprendre le taux d’acceptation**
+ Voici un exemple utilisant nos données de test. 

+ Si nous voulons accepter 85% de tous les prêts avec les probabilités de défaut les plus faibles, alors notre taux d’acceptation est de 85%. 

+ Cela signifie que nous rejetons 15% de tous les prêts présentant les probabilités de défaut les plus élevées. 

+ Au lieu de fixer une valeur seuil, nous voulons la calculer pour séparer les prêts que nous acceptons en utilisant notre taux d’acceptation des prêts que nous rejetons. Cette valeur ne sera pas la même que celle de 85 % que nous avons utilisée comme taux d’acceptation.

####**5. Calcul du seuil**
+ Pour calculer ce seuil, nous devons utiliser la fonction quantile de numpy. 

+ Cela utilise notre tableau de probabilités de défaut et le taux d’acceptation pour déterminer quelle valeur sépare nos prêts acceptés des prêts rejetés. Dans notre exemple, le seuil est de 0,804. 

+ Cela signifie que tous nos nouveaux prêts avec une probabilité de défaut inférieure à 80% sont acceptés, et toutes les probabilités supérieures sont rejetées.

####**7. Mauvais taux**
+ Même si nous avons calculé un taux d’acceptation pour nos prêts et fixé un seuil, il y aura toujours des défauts de paiement dans nos prêts acceptés. 

+ Ceux-ci sont souvent dans des plages de probabilité où notre modèle n’était pas bien calibré. 

+ Pour notre exemple, nous avons accepté 85% des prêts, mais tous ne sont pas en défaut de paiement comme nous le souhaiterions. 

+ Le mauvais taux est le pourcentage de prêts acceptés qui sont en fait des défauts de paiement. Donc, notre mauvais taux est un pourcentage des 10 016 prêts acceptés.

####**8. Mauvais calcul du taux**
+ Le calcul du mauvais taux est le nombre de défauts de paiement dans nos prêts acceptés divisé par le nombre total de prêts acceptés. 

+ En Python, nous utilisons la fonction somme de numpy pour compter le nombre de défauts et la méthode de comptage pour obtenir le nombre total de prêts acceptés. 

+ La raison pour laquelle cette méthode fonctionne est que nos valeurs non par défaut sont nulles et que les valeurs par défaut sont un, la somme est mathématiquement la même que le nombre de valeurs par défaut. Ceci divisé par le nombre de lignes dans le cadre de données nous donne notre mauvais taux!

+ Ce tableau est appelé tableau de stratégie parce que nous pouvons choisir notre taux d’acceptation en fonction de notre stratégie globale pour le portefeuille de prêts.

####**3. Mise en place de la table de stratégie**
+ Avant de commencer à calculer toutes les valeurs de la table de stratégie, nous configurons d’abord des objets pour stocker les valeurs. 

+ Ici, nous créons une liste de tous les taux d’acceptation à tester. Nous avons également deux listes vides pour stocker les seuils et les mauvais taux.

####**4. Calcul des valeurs de la table**

+ Pour créer la table, nous utilisons une boucle for pour effectuer tous les calculs automatiquement et stocker les résultats. 

+ Ainsi, pour chaque taux d’acceptation que nous définissons, nous calculons le seuil, le stockons pour plus tard, l’appliquons aux prêts, créons un sous-ensemble appelé prêts acceptés, puis calculons et stockons le mauvais taux.

####**5. Interprétation de la table de stratégie**
+ Une fois la boucle for terminée, nous pouvons créer un cadre de données pour le tableau de stratégie qui contient les valeurs de chacune des listes que nous avons faites auparavant. 

+ Si nous avons décidé d’un taux d’acceptation de 90%, nous pouvons voir ici le seuil et le mauvais taux de ce taux d’acceptation. 

+ Nous n’acceptons que 90% des nouveaux prêts, et rejetons les 10% les plus riches, ce qui nous donne un seuil de 0,947. 

+ Si nous supposons que nos prêts d’ensemble de test sont de nouveaux prêts jamais vus auparavant, nous avons un taux de mauvais paiement estimé à 13%.

####**6. Ajout des prêts acceptés**
+ Pour rendre le tableau de stratégie plus utile, nous allons ajouter quelques colonnes. 

+ Le premier est le nombre de prêts acceptés. Cela nous montre combien de nouveaux prêts nous acceptons pour chaque taux d’acceptation.

####**7. Ajout du montant moyen du prêt**
+ Ensuite, nous ajoutons une valeur moyenne du prêt que nous utiliserons plus tard. 

+ Il ne s’agit que du montant moyen du prêt de l’ensemble de test.

####**8. Estimation de la valeur du portefeuille**

+ Enfin, la valeur estimée est la différence entre la valeur moyenne des défauts acceptés sans défaut moins la valeur moyenne des défauts acceptés. 

+ Rappelez-vous, les défauts acceptés sont représentés sous forme de pourcentage qui est le mauvais taux. 

+ Ici, je suppose que chaque défaut de paiement dans nos prêts acceptés est une perte du montant moyen de tous les prêts. 

+ Cela donne une estimation approximative de la valeur du portefeuille à chaque taux d’acceptation. 

+ Donc, si je choisis un taux d’acceptation de 95%, mon mauvais taux est de 17,7%. Si mon taux d’acceptation est de 85%, mon mauvais taux est de 9%. 

+ Il y a un compromis entre l’acceptation et le mauvais taux. Avec ce nouveau tableau, nous pouvons estimer la différence.

####**9. Perte totale prévue**
+ La dernière façon dont nous mesurerons l’impact financier de nos prévisions est la perte totale attendue. 

+ Cela représente combien nous nous attendons à perdre sur les défauts de paiement compte tenu de leur probabilité de défaut. 

+ Nous prendrons le produit de la probabilité de défaut, de la perte en cas de défaut et de l’exposition au défaut pour chaque prêt, et nous l’additionnons. 

+ Dans le cadre de nos données de prévisions, nous utiliserons la probabilité de défaut de prob_default, l’exposition sera supposée comme la valeur totale du prêt et la perte en cas de défaut sera de 1 pour une perte totale sur le prêt.
